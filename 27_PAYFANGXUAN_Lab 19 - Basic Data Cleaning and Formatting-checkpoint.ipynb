## Part 1: Importing and Inspecting the Dataset

### Step 1: Import Required Libraries

First, we need to import the necessary libraries for data manipulation and analysis.


```python
# Import all necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

<h3>Step 2: Load the Dataset</h3>
<p>We will use the Titanic dataset for our data cleaning exercises.</p>

```python
# Load the Titanic dataset from a URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Display the first few rows of the dataset
df.head(10)
```

<h3>Step 3: Inspect the Dataset</h3>
<p>We will inspect the dataset to check for missing values, duplicates, and basic formatting issues.</p>

```python
# Get basic information about the dataset
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

<h3 id="Step-4:-Visualizing-Missing-Data">Step 4: Visualizing Missing Data</h3>
<p>Use visualization techniques to identify and understand the pattern of missing data.</p>

```python
# Visualize missing data using a heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(),cbar=False,cmap='magma')
plt.title("Missing Data Heatmap")
plt.show()
```

---

<h2>Part 2: Handling Missing Data</h2>
<h3>Step 1: Identify Missing Values</h3>
<p>Missing data is a common issue in real-world datasets. We need to identify and decide how to handle these missing values.</p>

```python
# Calculate the percentage of missing values per column
missing_percentage = df.isnull().sum()/len(df)*100
missing_percentage
```

<h3>Step 2: Filling Missing Values</h3>
<p>For this lab, we will fill the missing values in the <strong>Age</strong> column with the median age, and the <strong>Embarked</strong> column with the mode (most frequent value).</p>

```python
# Fill missing values in 'Age' with the median age
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
# Fill missing values in 'Embarked' with the most frequent value (mode)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
# Check if the missing values have been filled
df.isnull().sum()
```

<h3>Step 3: Removing Columns with High Missing Values</h3>
<p>In some cases, a column might have too many missing values to be useful. We will remove the <strong>Cabin</strong> column since it has a high percentage of missing values.</p>

```python
# Drop the 'Cabin' column due to excessive missing values
df.drop(columns=['Cabin'],inplace = True)
```

```python
# Verify the column has been removed
df.head(10)
```

---

<h2>Part 3: Removing Duplicate Records</h2>
<p>Duplicate records can cause bias in analysis, so it is important to check and remove duplicates.</p>

```python
# Check for duplicate rows
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")
```

```python
# Remove duplicate rows if any exist
df_cleaned = df.drop_duplicates()
```

```python
# Verify that duplicates have been removed
df_cleaned.shape
```

---

<h2>Part 4: Basic Data Formatting</h2>
<h3>Step 1: Renaming Columns</h3>
<p>Clear and descriptive column names improve the readability of the dataset. We will rename some columns for clarity.</p>

```python
# Rename columns for better understanding
df_cleaned.rename(columns={
    'Pclass':'PassengerClass',
    'SibSp':'SiblingSpousesAboard',
    'Parch':'ParentsChildrenAboard',
    'Fare':'TicketFare'
},inplace=True)
```

```python
# Display the new column names
df_cleaned.columns
```

```python
# Verify the column name has been renamed
df.head(10)
```

<h3>Step 2: Converting Data Types</h3>
<p>Ensuring the correct data types for columns is crucial for accurate analysis. We will convert the <strong>Survived</strong> column to a Boolean type and <strong>PassengerClass</strong> to a categorical type.</p>

```python
# Original data types
df_cleaned.dtypes
```

```python
# Convert 'Survived' to boolean type (0: False, 1: True)
df_cleaned['Survived']=df_cleaned['Survived'].astype(bool)
```

```python
# Convert 'PassengerClass' to categorical data type
df_cleaned['PassengerClass'] = df_cleaned['PassengerClass'].astype('category')
```

```python
# Verify the data types
df_cleaned.dtypes
```

---

<h2>Part 5: Removing Unnecessary Columns</h2>
<p>Sometimes, certain columns may not be useful for analysis. We will remove the <strong>Name</strong> and <strong>Ticket</strong> columns as they are not relevant for most analyses.</p>

```python
# Drop unnecessary columns
df_cleaned.drop(columns=['Name','Ticket'],inplace=True)
```

```python
# Display the cleaned dataset
df_cleaned.head()
```

---

<h2>Part 6: Visualising the cleaned dataset</h2>
<p style="margin-top: 1em; font-size: 14px;">Use visualisation techniques to verify the cleaned dataset.</p>

```python
# Visualize missing data using a heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df_cleaned.isnull(), cbar=False,cmap='viridis')
plt.title("Missing Data Heatmap")
plt.show()
```

---

<h2>Part 7: Saving the Cleaned Dataset</h2>
<p>Once the dataset is cleaned and formatted, we will save it as a CSV file for future use.</p>

```python
# Save the cleaned dataset to a new CSV file
df_cleaned.to_csv("titanic_cleaned.csv", index=False)
print("Cleaned dataset saved as 'titanic_cleaned.csv'")
```
