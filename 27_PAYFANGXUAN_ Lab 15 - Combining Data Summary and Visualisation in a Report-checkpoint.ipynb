## Step 1: Import Required Libraries
We will start by importing the necessary Python libraries for data analysis and visualisation.


```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set visualisation style
sns.set(style="whitegrid")
%matplotlib inline
```

---

<h2>Step 2: Load the Dataset</h2>
<p>For this lab, we will use the Titanic dataset from Seaborn. This dataset contains information about passengers on the Titanic, including their survival status, age, fare, and more.</p>

```python
# Load the Titanic dataset from Seaborn
df = sns.load_dataset('titanic')

# Display the first few rows of the dataset
df.head()
```

---

<h2>Step 3: Generate Summary Statistics for Numerical Columns</h2>
<p>We will first generate summary statistics for all numerical columns in the dataset, such as age, fare, etc.</p>

```python
# Generate summary statistics for numerical columns
summary_statistics = df.describe()
summary_statistics
```

<h2>Step 4: Generate Additional Summary Statistics for Categorical Data</h2>
<p>Next, we will generate summary statistics for categorical columns such as sex, class, and embarked. These statistics include counts and mode (most frequent values).</p>

```python
# Generate summary statistics for categorical columns
categorical_summary = df.describe(include=['object','category'])
categorical_summary
```

### Question:

---

<h2>Step 5: Visualise the Distribution of Numerical Variables</h2>
<p>We will now create histograms and KDE plots to visualise the distribution of key numerical variables such as age and fare.</p>
<h3>Distribution of Passenger Age</h3>

```python
# Plot histogram for age
plt.figure(figsize=(8,6))
sns.histplot(df['age'].dropna(),bins=20,kde=False,color='skyblue')
plt.title('Distribution of Passenger Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

<h3>KDE Plot of Passenger Fare</h3>

```python
# Plot KDE for fare
plt.figure(figsize=(8,6))
sns.kdeplot(df['fare'],shade=True,color='orange')
plt.title('KDE Plot of Passenger Fare')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.show()
```

### Question:
#### Question 1: Based on the age distribution, what is the most common age range of passengers?
- Based on the age distribution shown in the histogram, the most common age range of passengers is approximately 20 to 30 years.

#### Question 2: What can you infer about the distribution of fares from the KDE plot?
From the KDE plot of passenger fares, we can infer the following:

1. **Right-Skewed Distribution**: The distribution is heavily right-skewed, meaning most passengers paid lower fares, with fewer passengers paying very high fares.
2. **Peak at Lower Fares**: The highest density of fares is concentrated around the lower end, indicating that the majority of passengers paid fares close to this lower range.
3. **Long Tail**: There is a long tail extending towards higher fares, showing that while most fares were low, a small number of passengers paid significantly higher amounts.

This skewness could be due to different class accommodations (e.g., third-class fares being much lower than first-class fares).

---

<h2>Step 6: Visualise Categorical Data Using Bar Plots</h2>
<p>We will create bar plots to visualise the distribution of categorical variables such as sex and class.</p>
<h3>Count of Passengers by Sex</h3>

```python
# Bar plot for sex distribution
plt.figure(figsize=(8,6))
sns.countplot(x='sex',data=df,palette='pastel')
plt.title('Count of Passengers by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

<h3>Count of Passengers by Class</h3>

```python
# Bar plot for class distribution
plt.figure(figsize=(8,6))
sns.countplot(x='class',data=df,palette='muted')
plt.title('Count of Passengers by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

### Question:
#### Question 1: What does the bar plot tell you about the distribution of male and female passengers?
- The bar plot shows that there were significantly more male passengers than female passengers on the Titanic. Specifically, the count of male passengers is much higher than that of female passengers, indicating a gender imbalance among the passengers.

#### Question 2: What can you infer about the distribution of fares from the KDE plot?
The bar plot for passenger distribution across different classes shows:
* Third Class had the highest number of passengers.
* First Class and Second Class had significantly fewer passengers compared to Third Class.
* First Class had a slightly higher number of passengers than Second Class.

This indicates that the majority of passengers traveled in Third Class, with progressively fewer passengers in Second and First Classes.

<h2>Step 7: Visualise Relationships Between Variables</h2>
<p>We will now use scatter plots and pair plots to visualise the relationships between variables such as age, fare, and survived.</p>
<h3>Scatter Plot of Age vs Fare by Survival Status</h3>

```python
# Scatter plot for age vs fare, coloured by survival status
plt.figure(figsize=(10,8))
sns.scatterplot(x='age', y='fare', hue='survived',data=df,palette='coolwarm',alpha=0.6)
plt.title('Scatter Plot of Age vs Fare by Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', labels=['No','Yes'])
plt.show()
```

<h3>Pair Plot for Age, Fare, and Survived</h3>

```python
# Pair plot for age, fare, and survived
sns.pairplot(df[['age','fare','survived']],hue='survived',palette='Set2',diag_kind='kde')
plt.show()
```

### Question:
#### Question 1: What trends do you observe in the relationship between age, fare, and survival status?
From the scatter plot and pair plot, we can observe the following trends regarding the relationship between age, fare, and survival status:

1. **Fare and Survival**: Passengers who paid higher fares had a higher likelihood of survival. This may indicate that passengers in higher classes, who paid more, had better access to lifeboats or other safety measures.

2. **Age Distribution**: There is no strong direct correlation between age and survival. Passengers of various ages are distributed across both survival and non-survival categories. However, it seems that younger passengers are slightly more likely to survive than older ones.

3. **Clusters**: There are noticeable clusters of passengers paying low fares with mixed survival outcomes, but higher fares show more concentrated survival.

4. **Fare Outliers**: A few outliers exist with extremely high fares, and most of these passengers survived.

These trends highlight that class and fare may have played a significant role in survival chances, while age was less directly influential.

#### Question 2: Are there any noticeable patterns between age and fare for passengers who survived vs. those who did not?
Yes, there are some noticeable patterns between age and fare for passengers who survived versus those who did not:

1. **Fare Patterns**:
	- Survivors: Passengers who survived tend to have a higher concentration in the higher fare ranges. This suggests that those who could afford more expensive tickets, often in higher classes, had better survival rates.
	- Non-Survivors: A significant portion of non-survivors paid lower fares, indicating they were likely in lower classes, which had fewer survival advantages.

2. **Age Patterns**:
	- There is no strong age-based distinction between survivors and non-survivors, as both groups include passengers of various ages.
	- However, among younger passengers, there appears to be a slightly higher survival rate, although this pattern is less pronounced than the fare-related trends.

Overall, fare (and by extension, class) seems to have a stronger correlation with survival than age.
