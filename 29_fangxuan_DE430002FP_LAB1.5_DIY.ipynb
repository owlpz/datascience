# Set the dataframe index using set_index()

<p><span style="font-size: 12pt; color: #236fa1;">In pandas dataframes, we can set the index (row labels) using one or more existing columns or arrays (of the correct length). </span></p>
<p><span style="font-size: 12pt; color: #236fa1;">The index can replace the existing index or expand on it.</span></p>

```python
# create student_record for demonstration
student_record = {
    'school_code': ['s001', 's002', 's003', 's001', 's002', 's004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco', 'Gino Mcneill', 'Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_Of_Birth': ['15/05/2002', '17/05/2002', '16/02/1999', '25/09/1998', '11/05/2002', '15/09/1997'],
    'weight': ["35", "32", "33", "30", "31", "32"],
    'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4'],
    't_id': ['t1', 't2', 't3', 't4', 't5', 't6']
}

# create dataframe from student_record
student_record_df = pd.DataFrame(student_record)

# display student_record_df
student_record_df
```

<div>

</div>

<p><span style="font-size: 12pt; color: #236fa1;">From the dataframe created, we observed that the row index is automatically created on the left side.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">We can set the row index using the existing column.</span></p>

<div>

</div>

```python
# use the 't_id' to set as row index
# we use "inplace = True" to make the dataframe modification inplace
student_record_df.set_index('t_id', inplace = True)

# display student_record_df
student_record_df
```

<p><span style="font-size: 12pt; color: #236fa1;">After setting the new row index, we notice the "t_id" column is moved to the left side of dataframe.</span></p>

<div>

</div>

# Reset the dataframe index using reset_index()

<p><span style="font-size: 12pt; color: #236fa1;">We can use an existing column to set as an index, we can also reset it.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">reset_index() method sets a list of integers ranging from 0 to length of data as index.</span></p>

```python
# reset the index of the dataframe, and use the default one instead
student_record_df.reset_index(inplace = True)

# display student_record_df
student_record_df
```

# Concatenate two dataframes using pd.concat()

<p><span style="font-size: 12pt; color: #236fa1;">Concatenation is a simpler way to combine dataframes. </span></p>
<p><span style="font-size: 12pt; color: #236fa1;">It&rsquo;s often used to form a single, larger set to do additional operations on.</span></p>

```python
# create two dataframe for demonstration
student_data1 = {
    'Name': { 0: 'Jun Meng', 1: 'Ali', 2: 'Muthu' },
    'Gender': { 0: 'Male', 1: 'Male', 2: 'Male' },
    'Race': { 0: 'Chinese', 1: 'Malay', 2: 'Indian' },
    'Age': { 0: 18, 1: 20, 2: 20 }}

student_data2 = { 
    'Name': { 0: 'Fatimah', 1: 'Joyce' },
    'Gender': { 0: 'Female', 1: 'Female' },
    'Race': { 0: 'Malay', 1: 'Other' },
    'Age': { 0: 21, 1: 18 }}

student_data1_df = pd.DataFrame(student_data1)
student_data2_df = pd.DataFrame(student_data2)
```

```python
# display student_data1_df
student_data1_df
```

```python
# display student_data2_df
student_data2_df
```

<p><span style="font-size: 12pt; color: #236fa1;">We create two dataframes above for demonstration purposes.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">In the code below, we will concatenate two dataframes as one.</span></p>

```python
# use pd.concat() to concatenate two dataframes
combine_data_df = pd.concat([student_data1_df, student_data2_df])

# display combine_data1_df
combine_data1_df
```

<p><span style="font-size: 12pt; color: #236fa1;">Now we see the result after concatenating two dataframes together.</span></p>
