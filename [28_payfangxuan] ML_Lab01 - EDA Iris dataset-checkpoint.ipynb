<p><span style="color: #236fa1;">Unique values in the categorical data column</span></p>

---

<p><span style="text-decoration: underline; color: #236fa1;"><strong>Reviewing categorical data distribution</strong></span></p>
<p><span style="color: #236fa1;">In the statistical summary above, you notice the "Species" column was not listed. The "Species" column is categorical data and it is classified as a string. That's why it was not calculated and listed in the statistical summary.</span></p>
<p><span style="color: #236fa1;">Categorical data distribution statistics are useful in classification problems where we need to know the balance of categorical data.</span></p>
<p><span style="color: #236fa1;">In case we have an imbalanced categorical data distribution (one class is having lots more records than another class). It may need special handling at data preparation stage of the machine learning project.</span></p>
<p><span style="color: #236fa1;">In Python, we use the groupby(), size(), reset_index() method to find out the categorical data distribution.</span></p>

```python
# find out the distribution 
species_dist = iris_df.groupby("Species").size().reset_index(name = 'count')
species_dist
```

---

<p><span style="text-decoration: underline; color: #236fa1;"><strong>Visualise the distribution</strong></span></p>
<p><span style="color: #236fa1;">We can use direct use the Pandas plot to plot the categorical data distribution.</span></p>
<p><span style="color: #236fa1;">A bar plot is a plot that presents categorical data with rectangular bars with lengths proportional to the values that they represent.</span></p>
<p><span style="color: #236fa1;">A bar plot shows comparisons among discrete categories. One axis of the plot shows the specific categories being compared, and the other axis represents a measured value.</span></p>
<p>&nbsp;</p>
<p><span style="color: #236fa1;">Source: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.bar.html</span></p>

```python
# using bar chart, plot the distribution of categorical data
species_dist.plot.bar(title = 'Species distribution',
                     ylabel = 'Count',
                     xlabel = 'Species',
                     figsize = (6,3))
```

---

<p><span style="text-decoration: underline; color: #236fa1;"><strong>Using Seaborn to visualise the distribution</strong></span></p>
<p><span style="color: #236fa1;">Alternatively, we can use Seabron to visualise the categorical data distribution.</span></p>

```python
# alternative, use seaborn to plot the same distribution of categorical data
sns.barplot(x = "Species", y="count", data = species_dist)
```

---

<p><span style="text-decoration: underline; color: #236fa1;"><strong>Categorical data encoding</strong></span></p>
<p><span style="color: #236fa1;">We need to convert categorical data into an integer. One of the methods is easily applied one-hot encoding to this problem but that would simply increase the number of features and required higher computation power.</span></p>
<p><span style="color: #236fa1;">Instead, we derive a relation and map those categorical data.</span></p>
<p><span style="color: #236fa1;">In Python, we use map() method to map those categorical data to integer.</span></p>

```python
# find out the unique values in the "Species" column
unique_value = iris_df["Species"].unique()
display(unique_value)

# map the unique values in the "Species" column to integer
iris_df['Species'] = iris_df['Species'].map({unique_value[0]:1,
                                            unique_value[1]:2,
                                            unique_value[2]:3,})

# display the records after mapping the categorical data to integer
# you will observe that Iris-setosa mapped to 1, Iris-versicolor mapped to 2, 
# Iris-virginica mapped to 3
display(iris_df[45:55])
display(iris_df[95:105])
```

---

<p><span style="text-decoration: underline; color: #236fa1;"><strong>Reviewing correlation between columns</strong></span></p>
<p><span style="color: #236fa1;">The relationship between two variables is called correlation.</span></p>
<p><span style="color: #236fa1;">In statistics, the most common method for calculating correlation is Pearson&rsquo;s Correlation Coefficient. </span><br /><span style="color: #236fa1;">It can have three values as follows:</span></p>
<p style="padding-left: 40px;"><span style="color: #236fa1;">1. <strong>Coefficient value = 1</strong>: It represents the <span style="text-decoration: underline;">strong positive correlation</span> between variables.</span></p>
<p style="padding-left: 40px;"><span style="color: #236fa1;">2. <strong>Coefficient value = -1</strong>: It represents the <span style="text-decoration: underline;">strong negative correlation</span> between variables.</span></p>
<p style="padding-left: 40px;"><span style="color: #236fa1;">3. <strong>Coefficient value = 0</strong>: It represents <span style="text-decoration: underline;">no correlation</span> at all between variables.</span></p>
<p><span style="color: #236fa1;">It is always good to review the pairwise correlations of the variables in the dataframe before using them in the machine learning project because some machine learning algorithms such as linear regression and logistic regression will perform poorly if we have highly correlated variables.</span></p>
<p><span style="color: #236fa1;">In Python, corr() method is used to calculate a correlation matrix of dataframe.</span></p>

```python
# find the pearson's correlation coefficient of the dataframe
iris_df.corr(method='pearson')
```

---

<p><span style="text-decoration: underline; color: #236fa1;"><strong>Visualise the correlation of variables using the heatmap</strong></span></p>
<p><span style="color: #236fa1;">We can visualise the correlation of variables using Seaborn heatmap. Heatmap plot rectangular data as a colour-encoded matrix.</span></p>

```python
# change the plotting style
sns.set(style = "ticks", color_codes = True)

# use seaborn heatmap to plot the correlation coefficient
sns.heatmap(iris_df.corr(method='pearson'),center=0,annot=True)
```

<p><span style="text-decoration: underline; color: #236fa1;"><strong>Drop columns from dataframe</strong></span></p>
<p><span style="color: #236fa1;">In the dataframe, we notice the "Id" column is not relevant. It can be removed completely.</span></p>
<p><span style="color: #236fa1;">To drop specified labels from rows or columns, we can use drop() method.</span></p>
<p><span style="color: #236fa1;">Remove rows or columns by specifying label names and corresponding axis, or by specifying directly index or column names.&nbsp;</span></p>

```python
# drop the unwanted column
iris_df = iris_df.drop("Id", axis=1)
iris_df
```

---

<p><span style="text-decoration: underline; color: #236fa1;"><strong>Visualise the relationship of the variables using pairplot</strong></span></p>
<p><span style="color: #236fa1;">Pairplot visualization comes in handy when we perform exploratory data analysis (EDA), it plot pairwise relationships in a dataset.</span></p>
<p><span style="color: #236fa1;">Pairplot visualizes given data to find the relationship between them where the variables can be numerical or categorical.</span></p>
<p><span style="color: #236fa1;">Pairplot is a module of seaborn library which provides a high-level interface for drawing attractive and informative statistical graphics.</span></p>

```python
# change the plotting style
sns.set(style = "ticks",color_codes = True)
# use seaborn to plot the pairplot
sns.pairplot(iris_df, hue = 'Species')
```

<p><span style="text-decoration: underline; color: #236fa1;"><strong>A bit more complicated pairplot</strong></span></p>
<p><span style="color: #236fa1;">We can add the linear regression line in each of the sub-plot.</span></p>

```python
# change the plotting style

sns.set(style = "ticks", color_codes = True)
# we can show the linear regression fit on the joint axes by passing kind="reg"
sns.pairplot(iris_df, kind = "reg", hue = 'Species')
```
