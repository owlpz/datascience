<h2>Step 1: Import Required Libraries</h2>
<p>We will begin by importing the necessary Python libraries for data manipulation and visualization.</p>

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style for better aesthetics
sns.set(style="ticks")
%matplotlib inline
```

<h2>Step 2: Load the Dataset</h2>
<p>For this lab, we will use the Iris dataset from the Seaborn library, which contains information about different species of iris flowers and various measurements.</p>

```python
# Load the Iris dataset
df = sns.load_dataset('iris')

# Display the first few rows of the dataset
df.head()
```

<h2>Step 3: Create a Simple Scatter Plot</h2>
<p>Scatter plots show the relationship between two continuous variables. Let&rsquo;s plot <code>sepal_length</code> against <code>sepal_width</code> to visualize the relationship between them.</p>

```python
# Create a scatter plot of sepal_length vs. sepal_width
plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal_length',y='sepal_width',data=df,hue='species',palette='deep')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

<h2>Step 4: Create a Scatter Plot for Different Variables</h2>
<p>Let&rsquo;s create another scatter plot, this time between <code>petal_length</code> and <code>petal_width</code>, to check if there&rsquo;s a relationship between these variables.</p>

<h2>Step 5: Visualise Multiple Variable Relationships with Pair Plots</h2>
<p>Pair plots allow you to visualize relationships between multiple pairs of variables simultaneously. Let&rsquo;s use a pair plot to examine the relationships among all numerical variables in the dataset.</p>

```python
# Create a pair plot for the Iris dataset
sns.pairplot(df,hue='species',palette='deep')
plt.suptitle('Pair Plot of Iris Dataset',y=1.02)
plt.show()
```

<h2>Step 6: Customising Pair Plots</h2>
<p>You can customize pair plots to display additional information. Let&rsquo;s add KDE (Kernel Density Estimation) plots on the diagonal to better visualize the distribution of each variable.</p>

```python
# Create a customised pair plot with KDE on the diagonal
sns.pairplot(df, hue='species',palette='deep',diag_kind='kde')
plt.suptitle('Customised Pair Plot with KDE',y=1.02)
plt.show()
```

<h2>Step 7: Exploring Subsets of Variables with Pair Plots</h2>
<p>Sometimes, it&rsquo;s useful to focus on specific pairs of variables. Let&rsquo;s create a pair plot for <code>sepal_length</code>, <code>sepal_width</code>, and <code>petal_length</code> to focus on these specific relationships.</p>

```python
# Create a pair plot with a subset of variables
sns.pairplot(df, vars=['sepal_length','sepal_width','petal_length'],hue='species',palette='deep')
plt.suptitle('Pair Plot of Selected Variables',y=1.02)
plt.show()
```

<h2>Additional Enhancements</h2>
<p>To further enhance your understanding and visualization skills, consider the following:</p>
<ul>
<li>
<p><strong>Annotate Specific Points:</strong> Highlight or annotate specific data points that are outliers or particularly interesting.</p>
</li>
</ul>

```python
# Example: Annotate the first data point
plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal_length',y='sepal_width',data=df, hue='species',palette='deep')
plt.title('Annonated Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
# Annotate the first data point
plt.annotate('First Point',
            xy=(df.sepal_length[0],df.sepal_width[0]),
            xytext=(df.sepal_length[0],df.sepal_width[0]+0.2),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.legend(title='Species')
plt.show()
```

<ul>
<li>

<ul>
<li>
<p><strong>Correlation Heatmap: Visualize the correlation matrix to quantify relationships.</p>
</li>
</ul>

```python
# Compute correlation matrix
corr = df.corr()

# Generate a heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm',fmt=".2f")
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()
```
