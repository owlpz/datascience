<h1 id="Introduction-of-linear-regression"><span style="color: #236fa1;">Introduction of linear regression</span></h1>

<p><span style="font-size: 12pt; color: #236fa1;">We will start with the most familiar linear regression, a straight-line fit to data. A straight-line fit is a model of the form</span></p>
<p style="padding-left: 80px;"><span style="font-size: 12pt; color: #236fa1;">y = mx+&nbsp; c</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Where,</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #236fa1;">1. m: slope of the line</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #236fa1;">2. c: y-intercept, where x is zero</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #236fa1;">3. x: independent variable</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #236fa1;">4. y: dependent variable</span></p>

<p><span style="text-decoration: underline; font-size: 12pt; color: #236fa1;"><strong>Linear regression explained</strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Linear regression uses the <strong>relationship between the data points</strong> to <strong>draw a straight line through all of them</strong>.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">This line can be <strong>used to predict future values</strong>.</span></p>
<p>&nbsp;</p>
<p><span style="font-size: 12pt; color: #236fa1;">In Python, it consists of <strong>many methods to find relationships between data points</strong> and <strong>draw a linear regression line</strong>.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">We will show you how to use these methods instead of going through the mathematical formula.</span></p>

<p><span style="font-size: 12pt; color: #236fa1;">In the example below, the <strong>x-axis represents age</strong>, and the <strong>y-axis represents speed</strong>.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">We have registered <strong>the kids' age and time taken to walk 50 meters recorded in seconds</strong>.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Let us see if the data we collected could be used in linear regression:</span></p>

```python
# data of kids' age and time taken to walk 50 meters recorded in seconds
age = np.array([6, 8, 9, 8, 2, 18, 3, 10, 5, 12, 13, 10, 7])
time_second = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

# change the size of the plot figure
plt.figure(figsize = (8,6))

# creates a scatter plot with age on the x-axis and time_second on the y-axis
plt.scatter(age, time_second)

# change title and labels for plot
plt.title("Time taken for kids walk 50 meters")
plt.xlabel("Age")
plt.ylabel("Time (seconds)")
# show the scatter plot
plt.show()
```

<p><span style="font-size: 12pt; color: #236fa1;">We can create dataframe using the two variables and visualise the correlation of variables using Seaborn heatmap.</span></p>

```python
# create data frame using the age and time_second
data_df = pd.DataFrame({"age":age,"time_seconds": time_second})

# display data_df first 5 records
data_df.head()
```

```python
# change the plotting style
sns.set(style = "ticks", color_codes = True, rc = {'figure.figsize':(5,5)})

# using seaborn heatmap to plot the heatmap of correlation
sns.heatmap(data_df.corr(method='pearson'),center = 0, annot=True)
```

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Note:</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">We observed a negative correlation coefficient of -0.78.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">We can use Scikit-Learn's LinearRegression estimator to fit this data and construct the best-fit line:</span></p>

```python
#importing the LinearRegression class from sklearn.linear_model
from sklearn.linear_model import LinearRegression

# create a new instance of the LinearRegression class and name it as lm
lm = LinearRegression()

# use the fit() method to train the linear regression model using age and time_second
# first parameter expected to be 2D array, we need to use age[:, np.newaxis]
# age[:, np.newaxis] will make the numpy array from 1 dimension to 2 dimension
lm.fit(age[:,np.newaxis],time_second)

# create a list of 20 points in xfit
# predict the yfit values for each point of xfit
xfit = np.linspace(0,20,21)
yfit = lm.predict(xfit[:, np.newaxis])
# display the xfit and yfit in dataframe
table_disp = pd.DataFrame({'age': xfit, "time predict": yfit})
table_disp
```

Now, we can plot the original data together with the predicted data

```python
# change the size of the plot figure
plt.figure(figsize = (8,6))

# creates a scatter plot with age on the x-axis and time_second on the y-axis
plt.scatter(age, time_second)

# creates a line plot with xfit on the x-axis and predicted yfit on the y-axis
plt.plot(xfit,yfit,color = 'red')

# change title and labels for plot
plt.title("Time taken for kids walk 50 meters\nFitted with predicted line")
plt.xlabel("Age")
plt.ylabel("Time(seconds)")
# show the plot
plt.show()
```

<h1 id="R^2---R-squared-value"><span style="color: #236fa1;">R<sup>2</sup> - R-squared value</span></h1>

<p><span style="font-size: 12pt; color: #236fa1;">R<sup>2</sup>, R-squared is a <strong>statistical measure of how close the data are to the fitted regression line</strong>. </span></p>
<p><span style="font-size: 12pt; color: #236fa1;">It is also known as the <strong>coefficient of determination</strong>, or the coefficient of multiple determination for multiple regression.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">The best possible score is 1.0, which can be negative (because the model can be arbitrarily worse). A constant model that always predicts the expected value of y, disregarding the input features, would get a score of 0.0.</span></p>
<p>&nbsp;</p>
<p><span style="font-size: 12pt; color: #236fa1;">The R-squared value is <strong>used to evaluate how the predicted line is well fitted</strong> to the data points.</span></p>

<h1 id="Calculate-R-squared:-Method-1"><span style="color: #236fa1;">[Method 1] - Calculate R-squared</span></h1>

<div>

</div>

<p><span style="font-size: 12pt; color: #236fa1;">Here, we will demonstrate how to <strong>manually calculate the R-squared value</strong>.</span></p>

```python
# predict the y dependent values using the original x independent values
y_pred = lm.predict(age[:, np.newaxis])

# create a dataframe of y-actual values and y-predicted values
pred_df = pd.DataFrame({"y_actual": time_second,"y_pred":y_pred})

# display pred_df
pred_df
```

```python
# calculate y-actual square value
pred_df["y_actual_sq"] = pred_df["y_actual"] ** 2

# calculate y-predicted square value
pred_df["y_pred_sq"] = pred_df["y_pred"] ** 2


# calculate y-actual * y-predicted value
pred_df["y_actual_sq * y_pred"] = pred_df["y_actual"] *pred_df["y_pred"]


# display pred_df
pred_df
```

<p><span style="font-size: 12pt; color: #236fa1;">Now, we calculated the sum of all values for each column and display them.</span></p>

```python
# print all necessary information
print("rows of dataframe: ", len(pred_df))
print("sum of y_actual: ", pred_df["y_actual"].sum())
print("sum of y_pred: ", pred_df["y_pred"].sum())
print("sum of y_actual_sq: ", pred_df["y_actual_sq"].sum())
print("sum of y_pred_sq: ", pred_df["y_pred_sq"].sum())
print("sum of y_actual * y_pred: ", (pred_df["y_actual"] * pred_df["y_pred"]).sum())
```

<p><span style="font-size: 12pt; color: #236fa1;">We can start to calculate the R-squared value.</span></p>

```python
# import the math module 
import math

# store necessary info
rows = len(pred_df)
X = pred_df["y_actual"].sum()
Y = pred_df["y_pred"].sum()
X2 = pred_df["y_actual_sq"].sum()
Y2 = pred_df["y_pred_sq"].sum()
X_Y = (pred_df["y_actual"] * pred_df["y_pred"]).sum()

# calculated R-squared value using formula
# R2 =  [ (nΣxy – (Σx)(Σy)) / (√nΣx2-(Σx)2 * √nΣy2-(Σy)2) ]2
calculated_R2 = ((rows * X_Y - X * Y) / (math.sqrt(rows * X2 - X ** 2)*
                                        math.sqrt(rows * Y2 - Y ** 2))) ** 2
calculated_R2
```

<h1 id="[Method-1]---Calculate-R-squared" style="font-family: 'Helvetica Neue', Helvetica, Arial, sans-serif; color: #000000;"><span style="color: #236fa1;">[Method 2] - Calculate R-squared</span></h1>

Source: https://4.bp.blogspot.com/-5b_VCO7o3Nw/WChE4FcaLGI/AAAAAAAAARw/SyubtH5Vf_8tITFwIKd8nCCYn5ws9AWsACLcB/s1600/05%2BR-squared.PNG
<div>

</div>

<p><span style="font-size: 12pt; color: #236fa1;">Where,</span></p>
<ul>
<li>
<p><span style="font-size: 12pt; color: #236fa1;"><strong>SS<sub>res</sub></strong> is the <strong>sum of squares of the residual errors</strong>.</span></p>
</li>
<li>
<p><span style="font-size: 12pt; color: #236fa1;"><strong>SS<sub>tot</sub></strong> is the <strong>total sum of the errors</strong>.</span></p>
</li>
</ul>

```python
# predict the y dependent values using the original x independent values
y_pred = lm.predict(age[:, np.newaxis])

# create a dataframe of y-actual values and y-predicted values
pred_df = pd.DataFrame({"Y_actual": time_second, "y_pred": y_pred})

# display pred_df
pred_df
```

```python
# calculate y-actual square value
pred_df["error1"] = pred_df["y_actual"] - pred_df["y_pred"]

# calculate y-predicted square value
pred_df["error1_sq"] = pred_df["error1"] ** 2

# calculate y-actual square value
pred_df["error2"] = pred_df["y_actual"] - pred_df["y_actual"].mean()

# calculate y-predicted square value
pred_df["error2_sq"] = pred_df["error2"] ** 2
pred_df
```

```python
# calculated R-squared value using formula
# R2 = 1- SSres / SStot
calculated_R2 = 1 - (pred_df["error_sq"].sum()/ pred_df["error2_sq"].sum())
calculated_R2
```

<h1 id="[Method-2]---Calculate-R-squared" style="font-family: 'Helvetica Neue', Helvetica, Arial, sans-serif; color: #000000;"><span style="color: #236fa1;">[Method 3] - Calculate R-squared</span></h1>

<p><span style="color: #236fa1;"><span style="font-size: 12pt;">The easiest method is using sklearn's metrics to calculate the R-squared value.</span></span></p>

```python
# import metrics from sklearn package
from sklearn import metrics

# predict the y dependent values using the original x independent values
y_pred = lm.predict(age[:, np.newaxis])

# calculate the R-squared value using metrics.r2_score() function
metrics.r2_score(time_second,y_pred)
```

<h1 id="slope-coefficient-and-intercept-of-the-predicted-data"><span style="color: #236fa1;">Slope coefficient and intercept of the predicted data</span></h1>

<p><span style="font-size: 12pt; color: #236fa1;">The <strong>slope coefficient</strong> and <strong>intercept</strong> of the predicted data are contained in the lm model's fit parameters, which in Scikit-Learn are always marked by a trailing underscore.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">We can display it using <strong>coef_</strong> and <strong>intercept_</strong>.</span></p>

```python
# retrieve and print the slope coefficient and intercept of the predicted data
print ("Linear model slope:   ",lm.coef_[0])
print("Linear Model intercept: ", lm.intercept_)
```

<h1 id="Plot-the-predicted-line-with-info"><span style="color: #236fa1;">Plot the predicted line with info</span></h1>

<p><span style="font-size: 12pt; color: #236fa1;">We can add the slope coefficient and intercept information display on the plot.</span></p>

```python
# change the size of the plot figure
plt.figure(figsize = (10,8))
# creates a scatter plot with age on the x-axis and time_second on the y-axis
plt.scatter(age, time_second)

# creates a line plot with xfit on the x-axis and predicted yfit on the y-axis
plt.plot(xfit,yfit,color = 'red')

# change title and labels for plot
plt.title("Time taken for kids walk 50 meters\nFitted with predicted line")
plt.xlabel("Age")
plt.ylabel("Time(seconds)")
# add a marker of the intercept point
plt.plot(0, lm.intercept_, ls = "", marker = "p", label = "points", color = 'orchid', markersize =10)

# add text to display intercept
text1 = 'Y-intercept: ' + "{:.2f}".format(lm.intercept_)
plt.annotate(
    text1,
    xy=(0, lm.intercept_), xytext=(4, 110),
    arrowprops=dict(facecolor='black', shrink=0.05)
)

# add text to display slope coefficient
text2 = 'Slope coefficient: ' + "{:.2f}".format(lm.coef_[0])
plt.annotate(
    text2,
    xy=(xfit[8], yfit[8]), xytext=(12, 100),
    arrowprops=dict(facecolor='black', shrink=0.05)
)

# show the plot
plt.show()
```
