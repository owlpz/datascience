<h1><span style="color: #e67e23;">Aggregation - Statistical analytics</span></h1>

<p><span style="font-size: 12pt; color: #e67e23;">We use aggregation functions to do some computation (min, max, mean, median) on the life expectancy column "lifeExp".</span></p>

```python
# minimum of column "lifeExp"
minimum = gapminder_df["lifeExp"].min()

# maximum of column "lifeExp"
maximum = gapminder_df["lifeExp"].max()

# mean of column "lifeExp"
mean = gapminder_df["lifeExp"].mean()

# median of column "lifeExp"
median = gapminder_df["lifeExp"].median()

# print all the aggregation values
print ('minimum of column "lifeExp": is', minimum)
print ('maximum of column "lifeExp": is', maximum)
print ('mean of column "lifeExp": is', mean)
print( 'median  of column "lifeExp": is', median)
```

<p><span style="font-size: 12pt; color: #e67e23;">We also can use agg() function to find the same values in column "lifeExp".</span></p>

```python
# use agg() to find minimum, maximum, mean and median of column "lifeExp"
gapminder_df["lifeExp"].agg([np.min, np.max, np.mean, np.median])
```

---

<h1 id="GroupBy-+-Aggregation"><span style="color: #e67e23;">Analytics further using GroupBy + Aggregation</span></h1>

<p><span style="font-size: 12pt; color: #e67e23;">We can compute the mean life expectancy of each continent using groupby() and aggregation function.</span></p>
<p><span style="font-size: 12pt; color: #e67e23;">Refer to example code below:</span></p>

```python
# compute mean of life expectancy for each continent
grouped_continent_mean = gapminder_df[['continent','lifeExp']].groupby('continent').mean()

# display grouped_continent_mean
grouped_continent_mean
```

<h1><span style="color: #e67e23;">DIY Assignment:</span></h1>

<p><span style="font-size: 12pt; color: #e67e23;">Q1. Continue using the gapminder_df dataframe above, and <span style="text-decoration: underline;"><strong>compute each country's mean life expectancy</strong></span>.</span></p>

```python
# Q1
# write python code here
grouped_country_mean = gapminder_df[['country','lifeExp']].groupby('country').mean()
grouped_country_mean
```

```python
# Q2
# write python code here
selected_country = ["Singapore", "Malaysia", "Indonesia", "Thailand", "Myanmar"]

selected_grouped_country_mean = grouped_country_mean.loc[selected_country]
selected_grouped_country_mean
```
