<h2>Part 2: Identifying and Handling Outliers</h2>
<h3>Step 1: Identify Outliers</h3>
<p>Outliers can skew data and negatively impact model performance. We will use the Interquartile Range (IQR) method to detect outliers in the <code>price</code> and <code>carat</code> columns.</p>
<h4>Visualize the Distribution</h4>

```python
# Plot histograms for 'price' and 'carat'
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.histplot(df['price'], kde=True)
plt.title('Distribution of Price')
plt.subplot(1,2,2)
sns.histplot(df['carat'], kde=True)
plt.title('Distribution of Carat')
plt.tight_layout()
plt.show()
```

<p><strong>Boxplots to Detect Outliers</strong></p>

```python
# Boxplots for 'price' and 'carat'
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.boxplot(y=df['price'])
plt.title('Boxplot of Price')
plt.subplot(1,2,2)
sns.boxplot(y=df['carat'])
plt.title('Boxplot of Carat')
plt.tight_layout()
plt.show()
```

<h4 id="Important-note:">Important note:</h4>
<p>You will notice the outliers in the boxplot above.&nbsp;</p>
<p>To handle these outliers, we need to c<span style="font-size: 14px;">alculate IQR and identify outliers.</span></p>

```python
# Function to calculate IQR bounds
def calculate_iqr_bounds(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound

# Calculate bounds for 'price'
price_lower, price_upper = calculate_iqr_bounds('price')
print(f"Price lower bound: {price_lower}")
print(f"Price upper bound: {price_upper}")
# Calculate bounds for 'carat'
carat_lower, carat_upper = calculate_iqr_bounds('carat')
print(f"Carat lower bound: {carat_lower}")
print(f"Carat upper bound: {carat_upper}")
```

<p>We will visualise the Interquartile Range (IQR) bounds on the histograms for both the <code>'price'</code> and <code>'carat'</code> columns.</p>

```python
# Plot histograms for 'price' and 'carat'
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.histplot(df['price'],kde=True)
plt.title('Distribution of Price')

# Plot the histogram for 'price'
plt.axvline(price_lower,color='red', linestyle='--',linewidth=2,label='Lower IQR Bound')
plt.axvline(price_upper,color='red', linestyle='--',linewidth=2,label='Upper IQR Bound')

plt.fill_between(y=[0, plt.gca().get_ylim()[1]], x1=price_lower, x2=price_upper,
                  color='red', alpha=0.1,label='IQR Range')
plt.legend()
# Add vertical lines for IQR bounds


# Optional: Highlight the IQR range


# Plot the histogram for carat


# Add vertical lines for IQR bounds


# Optional: Highlight the IQR range
plt.subplot(1,2,1)
sns.histplot(df['carat'],kde=True)
plt.title('Distribution of Carat')

# Plot the histogram for 'price'
plt.axvline(carat_lower,color='purple', linestyle='--',linewidth=2,label='Lower IQR Bound')
plt.axvline(carat_upper,color='purple', linestyle='--',linewidth=2,label='Upper IQR Bound')

plt.fill_between(y=[0, plt.gca().get_ylim()[1]], x1=price_lower, x2=price_upper,
                  color='red', alpha=0.1,label='IQR Range')
plt.legend()
plt.tight_layout()
```

<h4>Important notes:</h4>
<ul>
<li>
<p><strong>Distribution of Price</strong>:</p>
<ul>
<li>A histogram showing the distribution of the <code>'price'</code> column.</li>
<li>Red dashed lines indicating the lower and upper IQR bounds.</li>
<li>A light red shaded area representing the IQR range.</li>
</ul>
</li>
<li>
<p><strong>Distribution of Carat</strong>:</p>
<ul>
<li>A histogram showing the distribution of the <code>'carat'</code> column.</li>
<li>Purple dashed lines indicating the lower and upper IQR bounds.</li>
<li>A light purple shaded area representing the IQR range.<br /><br /></li>
</ul>
</li>
</ul>
<p>These visual cues help in quickly identifying the spread of the data and potential outliers beyond the IQR bounds.</p>

<p>Identify the number of outliers</p>

```python
# Identify outliers in 'price'
price_outliers = df[(df['price'] < price_lower) | (df['price'] > price_upper)]
print(f"Number of outlier in 'price': {len(price_outliers)}")
# Identify outliers in 'carat'
carat_outliers = df[(df['price'] < carat_lower) | (df['carat'] > carat_upper)]
print(f"Number of outlier in 'carat': {len(carat_outliers)}")
```

<h3>Step 2: Handling Outliers</h3>
<p>We can choose to remove or cap outliers. Here, we will cap the outliers to the upper and lower bounds defined by the IQR.</p>

```python
# Cap outliers in 'price' and 'carat'
df['price_capped'] = np.where(df['price']>price_upper,price_upper,
                             np.where(df['price']< price_lower, price_lower, df['price']))

df['carat_capped'] = np.where(df['carat']>carat_upper,carat_upper,
                             np.where(df['carat']< carat_lower, carat_lower, df['carat']))
```

Boxplots after handling outliers

```python
# Visualize after capping
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.boxplot(y=df['price_capped'])
plt.title('Boxplot of Price after Capping')

plt.subplot(1,2,2)
sns.boxplot(y=df['carat_capped'])
plt.title('Boxplot of Carat after Capping')

plt.tight_layout()
plt.show()
```

<h4>Important note:</h4>
<p>No outlier detected after capping the outlier with the IQR bound.</p>

<h3 id="Standardize-the-'cut'-Column">Standardize the <code>'cut'</code> Column</h3>
<p>To ensure consistency, we'll perform the following transformations on the <code>'cut'</code> column:</p>
<ol>
<li><strong>Strip Whitespace</strong>: Remove any leading or trailing spaces.</li>
<li><strong>Convert to Lowercase</strong>: Standardize the text to lowercase.</li>
<li><strong>Replace Spaces with Underscores</strong>: Facilitate easier mapping by replacing spaces with underscores.</li>
</ol>

```python
# Standardize the 'cut' column: strip whitespace, convert to lowercase, and replace spaces with underscores
df['cut']=df['cut'].str.strip().str.lower().str.replace(' ', '_')

# Display the first few entries to verify changes
print("\n 'cut' column after standardization:")
print(df['cut'].unique())
```

<h4><strong>Code Explanation:</strong></h4>
<ul>
<li><strong><code>str.strip()</code></strong>: Removes any leading or trailing whitespace from each entry.</li>
<li><strong><code>str.lower()</code></strong>: Converts all characters in the string to lowercase, ensuring uniformity.</li>
<li><strong><code>str.replace(' ', '_')</code></strong>: Replaces spaces with underscores to handle multi-word categories like <code>'Very Good'</code>.</li>
<li>The <code>print</code> statement displays the first few entries of the <code>'cut'</code> column after these transformations to verify the changes.</li>
</ul>

<h3>Map Standardized Values to Proper Labels</h3>
<p>After standardization, we'll map the cleaned values back to their proper categorical labels. This step ensures that the categories are correctly named and consistent throughout the dataset.</p>

```python
# Define a mapping dictionary to convert standardized values to proper labels
cut_mapping = {
    'ideal': 'Ideal',
    'ideal': 'Ideal',
    'very good': 'Very Good',
    'verygood': 'Very Good',
    'premium': 'Premium',
    'premium_': 'Premium', 
    'good': 'Good',
    'fair': 'Fair',
    'fair_': 'Fair',
    '_fair_': 'Fair', 
    
}
# Apply the mapping to the 'cut' column 
df['cut'] = df['cut'].map(cut_mapping)
# Display unique values after mapping 
print("\nUnique 'cut' values after mapping:")
print(df['cut'].unique())
```

<p><strong>Code Explanation:</strong></p>
<ul>

<ul>
<li>Maps standardized lowercase and underscore-separated values back to their proper, formatted labels.</li>
<li>Includes an entry for <code>'ideal_'</code> to handle any unexpected trailing underscores that might exist after standardization.</li>
</ul>
</li>
<li><strong><code>df['cut'].map(cut_mapping)</code></strong>:
<ul>
<li>Applies the mapping to each entry in the <code>'cut'</code> column.</li>
<li>Transforms standardized strings to their intended categorical labels.</li>
</ul>
</li>
<li><strong>Handling Missing Mappings</strong>:
<ul>
<li>If there are entries in the <code>'cut'</code> column that do not match any key in <code>cut_mapping</code>, they will be converted to <code>NaN</code>.</li>

<h3><strong>Step 2.1: Forward-Fill and Backward-Fill</strong></h3>
<p><strong>Rationale:</strong></p>
<p>Forward-fill (<code>ffill</code>) and backward-fill (<code>bfill</code>) are simple imputation techniques that propagate the last valid observation forward or the next valid observation backward, respectively. They are especially useful for time-series data where the temporal order matters.</p>
<p><strong>Implementation:</strong></p>
<p>We'll apply these methods to the <code>'table'</code> column as an example.</p>

```python
# Check the number of missing values in 'table'
missing_table = df['table'].isnull().sum()
print(f"Number of missing values in 'table': {missing_table}")
```

<h3><strong>Step 2.4: Model-Based Imputation Using K-Nearest Neighbors (KNN)</strong></h3>
<p><strong>Rationale:</strong></p>
<p>Model-based imputation leverages machine learning algorithms to predict missing values based on other features. K-Nearest Neighbors (KNN) is a popular choice as it imputes missing values using the mean value from the nearest neighbors in the feature space.</p>
<p><strong>Implementation:</strong></p>
<p>We'll use <code>KNNImputer</code> from <code>scikit-learn</code> to perform imputation on the <code>'table'</code> column.</p>

<h4><strong>2.4.2: Prepare the Data for KNN Imputation</strong></h4>

```python
# Select numerical columns for KNN imputation
numeric_cols = ['carat_capped', 'depth_corrected','table','price_capped','x','y','z']

# Create a copy of the selected data
df_knn= df[numeric_cols].copy()

# Display the first few rows
print("\nSelected Numerical Columns for KNN Imputation:")
display(df_knn.head())
```

<p><strong>Code Explanation:</strong></p>
<ul>
<li><strong><code>numeric_cols</code></strong>: List of numerical features used for imputing <code>'table'</code>.</li>

<h4>2.4.3: Initialize and Apply KNN Imputer</h4>

```python
# Initialize KNN Imputer with 5 neighbors 
knn_imputer = KNNImputer(n_neighbors=5)
# Perform KNN imputation
df_knn_imputed = knn_imputer.fit_transform(df_knn)
# Convert the imputed array back to a dataframe
df_knn_imputed = pd.DataFrame(df_knn_imputed, columns=numeric_cols)
# Display the first few imputed rows
print("\nFirst Five Rows After KNN Imputation:")
display(df_knn_imputed.head())
```

<p><strong>Code Explanation:</strong></p>
<ul>
<li><strong><code>KNNImputer(n_neighbors=5)</code></strong>: Initializes the KNN imputer to consider the 5 nearest neighbors when imputing missing values.</li>
<li><strong><code>fit_transform(df_knn)</code></strong>: Fits the imputer on the data and transforms it by imputing missing values.</li>
<li><strong><code>pd.DataFrame(...)</code></strong>: Converts the imputed NumPy array back to a pandas dataframe with appropriate column names.</li>
</ul>
