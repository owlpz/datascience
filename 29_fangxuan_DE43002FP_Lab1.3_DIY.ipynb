# Data mapping using map()

<p><span style="color: #e67e23;"><span style="font-size: 16px;">We extract data from student_pc_internet.csv as an example.</span></span></p>

```python
# import pandas package
import pandas as pd 
```

```python
# read the data from "student_pc_internet.csv" file and store it as student_info_df
student_info_df = pd.read_csv("./student_pc_internet.csv")

# remove the "id" column
student_info_df.pop("id")

# display student_info_df
display(student_info_df)
```

<p><span style="font-size: 12pt; color: #e67e23;">From the student_info_df dataframe above, we observed that the laptop and internet columns contain values 1 and 0.</span></p>
<p><span style="font-size: 12pt; color: #e67e23;">The value 1 and 0 denotes "TRUE" and "FALSE" respectively.</span></p>

<div>

</div>

### Map the laptop column

<p><span style="font-size: 12pt; color: #e67e23;">By using the mapping function we can add one more column to an existing dataframe.<br /></span></p>

```python
# create a dictionary of the key values we wish to map and the replacement
Mapping_TF_dict = {1: 'TRUE', 0: 'FALSE'}

# using the 'laptop' to do the mapping and store it into a new column name 'laptop_TF'
student_info_df['laptop_TF'] = student_info_df['laptop'].map(Mapping_TF_dict)

# display the first 10 records
student_info_df.head(10)
```

# Data mapping using apply() and lambda

<p><span style="font-size: 12pt; color: #e67e23;">We can also perform data mapping using apply() and lambda functions.</span></p>

<p><span style="font-size: 12pt; color: #e67e23;">A lambda function is a small anonymous function.</span></p>
<p><span style="font-size: 12pt; color: #e67e23;">A lambda function can take any number of arguments, but can only have one expression.</span></p>
<p><span style="font-size: 12pt; color: #e67e23;">Some example below:</span></p>

```python
# lambda function example 1

# create a lambda fucntion to add 10
add_ten_func = lambda number1 : number1 + 10

# pass in the paramater add_ten_func() lambda function

add_ten_func(15)
```

```python
# lambda function example 2

# create a lambda fucntion to add 10
addition_func = lambda number1, number2, number3 : number1 + number2 + number3

# pass 3 paramaters addition_func() lambda function
addition_func(15,20,30)
```

```python
# lambda function example 3

# create a lambda fucntion to add 10
check_even_func = lambda number1: True if (number1 % 2) == 0 else False

# pass 3 paramaters addition_func() lambda function
check_even_func(1623)
```

### use the lambda function to do the mapping

<p><span style="color: #e67e23; font-size: 16px;">In the example below, we want to map the 'internet' column.</span></p>

```python
# create a lambda function to map the true and false
mapping_func_TF = lambda x: 'TRUE' if x == 1 else 'FALSE'

# using the 'internet' to do the mapping and store it into a new column name 'internet_TF'
student_info_df['internet_TF'] = student_info_df['internet'].apply(mapping_func_TF)

# display the first 10 records
student_info_df.head
```

# Create new datafram by selecting columns from existing dataframe

<p><span style="font-size: 12pt; color: #e67e23;">Once we done the mapping, we can create another dataframe by selecting the columns that we wanted from the student_info_df dataframe.</span></p>
<p><span style="font-size: 12pt; color: #e67e23;">To do this, we can specify the column name and pass in as a list, then use the copy() method.</span></p>
<p><span style="font-size: 12pt; color: #e67e23;">Example code below:</span></p>

```python
# specify the column name and pass in as a list
# use the copy() method to create a new dataframe name as new_student_info_df
new_student_info_df = student_info_df[['first_name','last_name','email','gender',
                                     'laptop_TF','internet_TF']].copy()


# display the first 10 records
new_student_info_df.head(10)
```

# Categorise the data

<p><span style="color: #e67e23;"><span style="font-size: 16px;">We extract data from tuition_class.csv as an example.</span></span></p>

```python
# read the data from "tuition_class.csv" file and store it as student_info_df
tuition_class_df = pd.read_csv("./tuition_class.csv")

# display tuition_class_df
display(tuition_class_df)
```

### Categorise data using cut() method

<p><span style="font-size: 12pt; color: #e67e23;">We will use cut when we need to segment and sort data values into bins (categorise the data). </span></p>
<p><span style="font-size: 12pt; color: #e67e23;">This function is also useful for going from a numerical variable to a categorical variable. </span></p>
<p><span style="font-size: 12pt; color: #e67e23;">For example, cut() could convert ages to groups of age ranges.&nbsp;</span></p>
<p><span style="font-size: 12pt; color: #e67e23;">In the example below, we categorise the income category based on the PCI.</span></p>

<div>

</div>

```python
# import numpy package
import numpy as np

# bins are the range of numbers to categorise the data
bins = [0, 650, 1050, np.inf]

# labels are the specifies the labels for the returned bins
names = ['low income', 'middle income', 'high income']

# using the pd.cut() to categorise the income
tuition_class_df['Income'] = pd.cut(tuition_class_df['PCI'],bins, labels = names)

# display tuition_class_df
display(tuition_class_df)
```

<p><span style="font-size: 12pt; color: #e67e23;">bins are the range of numbers to categorise the data.</span></p>
<p><span style="font-size: 12pt; color: #e67e23;">labels are the specifies the labels for the returned bins.</span></p>

<div>

</div>

### Categorise data using iterrows() & if-else

<p><span style="font-size: 12pt; color: #e67e23;">The DataFrame.iterrows() method iterates through Pandas DataFrame rows as (index, Series) pairs.</span></p>
<p><span style="font-size: 12pt; color: #e67e23;">The iterator returns a tuple of column names and data in the form of a series.</span></p>
<p><span style="font-size: 12pt; color: #e67e23;">In the example code below, we have a dataframe that contains four observations: "Name," "Age," "Gender," and "PCI"</span></p>

```python
# iterating over the datadrame rows Using df.iterrows()
for index, value in tuition_class_df.iterrows():
    print('index: ', index)
    print('value: ', value)
    print('-' * 50)


# code explaination:
# iterate through each row of the tuition_class_df dataframe
# print each observation as a tuple
# use iterrows() to get each observation as a tuple.
```

<p><span style="font-size: 12pt; color: #e67e23;">Now we already understand we can iterate the rows in a dataframe using interrows(). </span></p>
<p><span style="font-size: 12pt; color: #e67e23;">Combine with the if-else decision making, we can categorise the data according to specified condition. </span></p>
<p><span style="font-size: 12pt; color: #e67e23;">In example below, we categorise the "Age" to "under 18" and "18 &amp; above" as shown below.</span></p>

<div>

</div>

```python
# create a empty list name as age_limit
age_limit = list()

# iterate each row in the tuition_class_df dataframe
# check the age and categorise them into "under 18" and "18 & above"
for index, value in tuition_class_df.iterrows():
    if value['Age'] < 18:
        age_limit.append("under 18")
    else:
        age_limit.append("18 & above")


# create a new column using the age_limit list
tuition_class_df["Age_limit"] = age_limit

# display tuition_class_df
display(tuition_class_df)
```

### Find the data distribution using value_counts()

```python
# find out the "Age_limit" distribution/propotion
tuition_class_df['Age_limit'].value_counts()
```

```python
# find out the "Income" distribution/propotion
tuition_class_df['Income'].value_counts()
```
