## Step 1: Import Required Libraries

First, we need to import the necessary libraries for data manipulation, visualisation, and handling missing data.


```python
# Importing necessary libraries for data manipulation and visualisation
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer, KNNImputer

# Setting the style for seaborn plots for better aesthetics
sns.set(style="whitegrid")

# Ensuring that plots are displayed within the notebook
%matplotlib inline
```

---
## Step 2: Load Dataset

We'll use the Titanic dataset from seaborn's built-in datasets, which contains both numerical and categorical columns with missing values.


```python
# Loading the Titanic dataset from seaborn's built-in datasets
df = sns.load_dataset('titanic')

# Displaying the first few rows of the dataset to understand its structure
df.head()
```

---
## Step 3: Identifying Missing Data

### Concept
Missing data can appear as `NaN` (Not a Number) or empty cells in a dataset. Identifying where and how much data is missing is the first step to handling it effectively.


### Step 3.1: Checking for Missing Values in Each Column

Use the `.isnull()` method combined with `.sum()` to count the number of missing values in each column.


```python
# Printing the number of missing values in each column of the dataset
print("Missing values in each column:")
print(df.isnull().sum())
```

### Step 3.2: Visualising Missing Data

A heatmap can be useful to visualise where missing data is present.


```python
# Setting the figure size for the heatmap
plt.figure(figsize=(12,8))

# Creating a heatmap to visualise missing data
sns.heatmap(df.isnull(),cbar=False,cmap='viridis',yticklabels=False)

# Adding title and labels to the heatmap
plt.title("Heatmap of Missing Data")
plt.xlabel("Columns")
plt.ylabel("Rows")
# Displaying the heatmap
plt.show()
```

**Interpretation:**
- The `age` and `deck` columns have significant missing data.
- `embarked` and `embark_town` have minimal missing values.


---
## Step 4: Handling Missing Data

### Concept
There are several techniques to handle missing data:
1. **Removing rows** with missing data.
2. **Filling (imputing)** missing data with strategies like mean, median, or mode.
3. **Advanced imputation methods** (e.g., KNN imputation).


### Step 4.1: Removing Rows with Missing Data

Dropping rows that contain any missing values can lead to significant data loss, especially if missingness is widespread.


```python
# Dropping all rows that contain any missing values
df_dropped = df.dropna()

# Displaying the shape of the original dataset
print("Original dataset shape:",df.shape)

# Displaying the shape of the dataset after dropping missing data
print("New dataset shape after dropping missing data:",df_dropped.shape)
```

#### **Note:** Dropping rows can lead to the loss of valuable information, especially if the missing data is not random.

### Step 4.2: Imputing Missing Data (Using Mean, Median, Mode)

Imputing missing values helps retain the data while replacing missing entries with meaningful estimates.


#### Imputing Numerical Data (`age`)

We'll focus on the `age` column, which has missing data.


```python
# Printing the number of missing values in the 'age' column
print("Number of missing values in 'age':",df['age'].isnull().sum())
```

**Imputing with Mean:**


```python
# Creating a SimpleImputer object with strategy 'mean' to impute missing 'age' values
imputer_mean = SimpleImputer(strategy='mean')

# Applying the imputer to the 'age' column and creating a new column 'age_mean_filled' with 
# the imputed values
df['age_mean_filled']=imputer_mean.fit_transform(df[['age']])
```

**Imputing with Median:**


```python
# Creating a SimpleImputer object with strategy 'median' to impute missing 'age' values
imputer_median = SimpleImputer(strategy='median')

# Applying the imputer to the 'age' column and creating a new column 'age_median_filled' 
# with the imputed values
df['age_median_filled']=imputer_median.fit_transform(df[['age']])
```

**Imputing with Mode:**


```python
# Creating a SimpleImputer object with strategy 'most_frequent' (mode) to impute 
# missing 'age' values
imputer_mode = SimpleImputer(strategy='most_frequent')

# Applying the imputer to the 'age' column and creating a new column 'age_mode_filled' 
# with the imputed values
df['age_mode_filled'] = imputer_mode.fit_transform(df[['age']])
```

**Display the results:**


```python
# Displaying the first 10 rows of the original and imputed 'age' columns for comparison
df[['age','age_mean_filled','age_median_filled','age_mode_filled']].head(10)
```

---
### Step 4.3: Imputing Categorical Data

Categorical columns with missing values can be filled with the most frequent value (mode).


```python
# Printing the number of missing values in the 'embarked' column
print("Number of missing values in embarked:", df['embarked'].isnull().sum())
```

**Imputing `embarked` with Mode:**


```python
# Applying the mode imputation to the 'embarked' column and creating a new column 'embarked_filled'
df['embarked_filled'] = imputer_mode.fit_transform(df[['embarked']]).flatten()
```

**Display the results:**


```python
# Displaying the first 10 rows of the original and imputed 'embarked' columns for comparison
df[['embarked','embarked_filled']].head(10)
```

---

## Step 5: Visualising the Impact of Mean and Median Imputation

Understanding how imputation affects the data distribution is crucial. Let's visualise the distribution of the original and imputed `age` columns.


```python
# Setting the figure size for the histogram
plt.figure(figsize=(12,8))

# Plotting the distribution of the original 'age' column
sns.histplot(df['age'],kde=True,color='red',label='Original Age',alpha=0.6)

# Plotting the distribution of the 'age_mean_filled' column
sns.histplot(df['age_mean_filled'],kde=True,color='blue',label='Mean Imputed Age',alpha=0.6)

# Plotting the distribution of the 'age_median_filled' column
sns.histplot(df['age_median_filled'],kde=True,color='green',label='Median Imputed Age',alpha=0.6)

# Adding legend to differentiate between distributions
plt.legend()

# Adding title and labels to the histogram
plt.title("Impact of imputation on 'Age' Column")
plt.xlabel("Age")
plt.ylabel("Frequency")
# Displaying the histogram
plt.show()
```

### **Histogram Interpretation:**
- **Mean Imputation:** The distribution shifts towards the mean value, potentially reducing variance.
- **Median Imputation:** More robust to outliers, maintaining the central tendency better than mean.
- **Mode Imputation:** Not shown here but applicable for categorical data.

## Step 6: Discussion Questions

#### Q1: What are the potential risks of removing rows with missing data? When would this be an acceptable strategy?

**Answer:**
Removing rows with missing data can lead to significant data loss, especially if missingness is widespread. This can reduce the statistical power of the analysis and potentially introduce bias if the missingness is not completely random. However, it might be acceptable when the proportion of missing data is very small or when the missing data is randomly distributed and unlikely to bias the results.

#### Q2: Why might imputing missing data with the mean or median be problematic for certain datasets?

**Answer:**
Imputing with mean or median can distort the original data distribution, reduce variability, and potentially introduce bias, especially if the data is not normally distributed or has outliers. It also ignores the relationships between variables, potentially oversimplifying the data and leading to inaccurate models.

#### Q3: Can you think of more advanced methods for handling missing data beyond simple imputation?

**Answer:**
Advanced methods include:
- **K-Nearest Neighbors (KNN) Imputation:** Uses the values from the nearest neighbors to impute missing data.
- **Multiple Imputation:** Creates multiple complete datasets by imputing missing values multiple times and combines the results.
- **Regression Imputation:** Uses regression models to predict missing values based on other variables.
- **Machine Learning Algorithms:** Algorithms like Random Forests can handle missing data intrinsically.
- **Deep Learning Techniques:** Such as autoencoders for imputation.

<h3>Step 7: K-Nearest Neighbors (KNN) Imputation</h3>
<p><strong>KNN Imputation</strong> replaces missing values by considering the values of the <strong>k-nearest neighbors</strong> in the dataset. This method can capture more complex patterns in data, especially when features are correlated.</p>
<p><strong>Note:</strong> KNN Imputer requires numerical data. If your dataset contains categorical variables, you need to encode them or select only numerical columns.</p>
<p><strong>Step 7.1: Applying KNN Imputer</strong></p>

```python
# Select numerical columns for KNN Imputation
knn_columns = ['age','fare']

# Apply KNN Imputation to the selected columns
knn_imputer = KNNImputer(n_neighbors=5)
df_knn_imputed = knn_imputer.fit_transform(df[knn_columns])

# Add KNN-imputed 'age' and 'fare' columns back to the DataFrame
df['age_knn_imputed'] = df_knn_imputed[:,0]
df['fare_knn_imputed'] = df_knn_imputed[:,1]
```

<p><strong>Display the Results</strong></p>

```python
# Display the original and KNN-imputed 'age' and 'fare' columns
print("\nComparison of Original and KNN-Imputed 'age' and 'fare' Columns:")
print(df[['age','age_knn_imputed','fare','fare_knn_imputed']].head(10))
```

---

<h3>Step 8: Visualizing the Effect of Imputation</h3>
<p>Visualizing the distribution of original and imputed values can help assess the impact of imputation techniques on the data's distribution.</p>
<p><strong>Step 8.1: Plot Distributions of Original and Imputed 'Age' Columns</strong></p>

```python
# Plot the distributions of the original and imputed 'age' columns
plt.figure(figsize=(10,6))
sns.kdeplot(df['age'],color='red',label='Original Age',fill=True, alpha=0.3)
sns.kdeplot(df['age_mean_filled'], color='blue', label='Mean Imputed Age', fill=True, alpha=0.3)
sns.kdeplot(df['age_knn_imputed'], color='green', label='KNN Imputed Age', fill=True, alpha=0.3)
plt.legend()
plt.title("Effect of Imputation on 'Age' Column")
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```
