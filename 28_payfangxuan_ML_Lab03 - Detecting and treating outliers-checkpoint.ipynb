<h2 id="Detecting-outliers"><span style="color: #236fa1;">Detecting outliers</span></h2>
<p><span style="color: #236fa1; font-size: 12pt;">To identify the outliers in given datasets, we can use visualization and mathematical techniques:</span></p>
<ol>
<li><span style="font-size: 12pt; color: #236fa1;">Boxplot</span></li>
<li><span style="font-size: 12pt; color: #236fa1;">Inter Quantile Range (IQR)</span></li>
<li><span style="font-size: 12pt; color: #236fa1;">Z-score</span></li>
</ol>

---

<p><span style="text-decoration: underline;"><span style="color: #236fa1;"><span style="font-size: 16px;"><strong><u>[Detecting outliers] </u></strong></span></span></span><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Method 1: Detect outlier using boxplot</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be &ldquo;outliers&rdquo; using a method that is a function of the inter-quartile range.</span></p>

```python
#sns.set_theme(style="whitegrid")
sns.set(style = "ticks", color_codes = True, rc = {'figure.figsize':(8,3)})
sns.boxplot(x = diabetes_df["BloodPressure"])
```

---

<p><span style="text-decoration: underline; font-size: 12pt; color: #236fa1;"><strong>Understanding boxplots</strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;"><strong>Boxplots</strong> are a standardized way of displaying the distribution of data based on a <strong>five statistical summary</strong> (&ldquo;minimum&rdquo;, first quartile (Q1), median, third quartile (Q3), and &ldquo;maximum&rdquo;).</span></p>

Source: https://miro.medium.com/max/1400/1*2c21SkzJMf3frPXPAR_gZA.png
<div>

</div>

<p><span style="font-size: 12pt; color: #236fa1;"><strong>Median (Q2/50th Percentile):</strong> The middle value of the dataset.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;"><strong>First quartile (Q1/25th Percentile):</strong> The middle number between the smallest number (not the &ldquo;minimum&rdquo;) and the median of the dataset.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;"><strong>Third quartile (Q3/75th Percentile):</strong> The middle value between the median and the highest value (not the &ldquo;maximum&rdquo;) of the dataset.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;"><strong>Interquartile range (IQR):</strong> 25th to the 75th percentile.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;"><strong>Whiskers:</strong> Shown in blue</span></p>
<p><span style="font-size: 12pt; color: #236fa1;"><strong>Outliers:</strong> Shown as green circles</span></p>
<p><span style="font-size: 12pt; color: #236fa1;"><strong>&ldquo;Maximum&rdquo;:</strong> Q3 + 1.5 * IQR</span></p>
<p><span style="font-size: 12pt; color: #236fa1;"><strong>&ldquo;Minimum&rdquo;:</strong> Q1 - 1.5 * IQR</span></p>
<p>&nbsp;</p>
<p><span style="font-size: 12pt; color: #236fa1;">Source: https://towardsdatascience.com/understanding-boxplots-5e2df7bcbd51</span></p>

---

<p><span style="text-decoration: underline;"><span style="color: #236fa1;"><span style="font-size: 16px;"><strong><u>[Detecting outliers]&nbsp;</u></strong></span></span><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Method 2: </span></strong></span><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Detect outliers using the Inter Quantile Range (IQR)</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">From the boxplot explanation above, we know the following:</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">1. Maximum is 1.5 times IQR above Q3</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">2. Minimum is 1.5 times IQR below Q1</span></p>
<p>&nbsp;</p>
<p><span style="font-size: 12pt; color: #236fa1;">In the code below, we find out the Q1, Q3 and IQR first. And then we calculate the maximum and minimum values.<br /></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Once we know the maximum and minimum values, outliers are listed by finding out the records above maximum and below minimum.</span></p>

```python
# calculate the 25th and 75th percentile
q75, q25 = np.percentile(diabetes_df.loc[:,['BloodPressure']],[75,25])

# calculate the inter-quartile
intr_qr = q75 - q25

# calculate the 1.5 times IQR above Q3 and below Q1
boxplot_max = q75 + (1.5 * intr_qr)
boxplot_min = q25 - (1.5 * intr_qr)

# print the 1.5 times IQR above Q3 and below Q1
print('boxplot_max is', boxplot_max)
print('boxplot_min is', boxplot_min)
```

```python
# we find out and display all the records < boxplot_min
diabetes_df.loc[diabetes_df["BloodPressure"]< boxplot_min, "BloodPressure"]
```

```python
# we find out and display all the records > boxplot_max
diabetes_df.loc[diabetes_df["BloodPressure"]> boxplot_max,"BloodPressure"]
```

---

<p><span style="text-decoration: underline;"><span style="color: #236fa1;"><span style="font-size: 16px;"><strong><u>[Detecting outliers]&nbsp;</u></strong></span></span><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Method 3: </span></strong></span><span style="text-decoration: underline;"><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;"><span style="color: #236fa1;"><span style="font-size: 12pt;"><strong>Detect outliers using Z-score</strong></span></span></span></span><span style="font-size: 12pt; color: #236fa1;"><br /></span><span style="font-size: 12pt; color: #236fa1;"><span style="color: #236fa1;"><span style="font-size: 16px;">Z-score is an important concept in statistics. </span></span></span></p>
<p><span style="font-size: 12pt; color: #236fa1;"><span style="color: #236fa1;"><span style="font-size: 16px;">This score helps to understand if a data value is greater or smaller than the mean and how far away it is from the mean. </span></span></span></p>
<p><span style="font-size: 12pt; color: #236fa1;"><span style="color: #236fa1;"><span style="font-size: 16px;">More specifically, the Z-score tells how many standard deviations away a data point is from the mean.</span></span></span></p>

Source: https://www.inchcalculator.com/wp-content/uploads/2021/10/z-scores-for-normal-standard-distribution.png
<div>

</div>

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Z-score and Outliers</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">If the z score of a data point is more than 3, it indicates that the data point is quite different from the other data points. Such a data point can be an outlier.<br /></span></p>

Source: https://slideplayer.com/slide/12859452/78/images/57/Outlier+detection+with+z-score.jpg
<div>

</div>

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">How do we calculate z-score?</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">The formula for calculating a z-score is z = (x - &mu;) / &sigma;, where x is the raw score, &mu; is the population mean, and &sigma; is the population standard deviation.</span></p>

Source: https://toptipbio.com/wp-content/uploads/2020/02/Z-score-formula.jpg
<div>

</div>

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Calculate z-score using stats.zscore() function</span></strong></span></p>

```python
# import stats from scipy package
from scipy import stats

# compute the z score of each value in the sample, relative to the sample mean and standard deviation
# store the z-score in new column diabetes_df["z-score"]
diabetes_df["z-score"] = np.abs(stats.zscore(diabetes_df["BloodPressure"]))

# display diabetes_df
diabetes_df
```

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Display the outliers using z-score &gt; 3.0</span></strong></span></p>

```python
# display the records with z-score > 3.0, these are the outliers detected
diabetes_df.loc[diabetes_df["z-score"]> 3.0]
```

<h2 id="Treating-Outliers"><span style="color: #236fa1;">Treating Outliers</span></h2>
<p><span style="font-size: 12pt; color: #236fa1;">We learned about detecting the outliers, we can start to treat these outliers. Below are some of the methods of treating the outliers.</span></p>
<ol>
<li><span style="font-size: 12pt; color: #236fa1;">Trimming/removing the outlier</span></li>
<li><span style="font-size: 12pt; color: #236fa1;">Quantile based flooring and capping</span></li>
<li><span style="font-size: 12pt; color: #236fa1;">Mean/Median imputation</span></li>
</ol>

---

<p><span style="text-decoration: underline;"><span style="font-size: 14px;"><span style="color: #236fa1; text-decoration: underline;"><span style="font-size: 16px;"><strong>[Treating outliers] </strong></span></span></span><span style="font-size: 14px;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Method 1:&nbsp;</span></strong></span><span style="color: #236fa1;"><span style="font-size: 16px;"><strong>Trimming/removing the outlier using z-score</strong></span></span></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">In this technique, we remove the outliers from the dataset. Although it is not a good practice.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Python code to delete the outlier and copy the rest of the elements to another array.</span></p>

```python
# import stats from scipy package
from scipy import stats

# compute the z score of each value in the sample, relative to the sample mean and standard deviation
# filter records with z-score < 3.0
# copy into new dataframe diabetes_df_demo1
diabetes_df_demo1 = diabetes_df[(np.abs(stats.zscore(diabetes_df["BloodPressure"]))< 3)].copy()

# display diabetes_df_demo1
diabetes_df_demo1
```

```python
# find out the shape of the diabetes_df_demo1
# you will notice the records are lesser compare to the original dataframe
diabetes_df_demo1.shape
```

<p><span style="color: #236fa1; font-size: 16px;">We plot the KDE plot on the diabetes_df_demo1["BloodPressure"] column, you will notice the distribution looks symmetric and the skew value is closed to zero.</span></p>

```python
# change the plotting style and figure size
sns.set(style = "ticks", color_codes = True, rc = {'figure.figsize':(5,5)})

# plot the KDE with color and fill
sns.kdeplot(diabetes_df_demo1["BloodPressure"], color = "brown", fill = True)

# find out the skewness of variable
skewness_0 = diabetes_df_demo1["BloodPressure"].skew()

# display the skewness value by add text annotation on plot
plt.text(40, 0.036, "Skew:" + str(skewness_0), size=16)
```

---

<p><span style="text-decoration: underline;"><span style="font-size: 14px;"><span style="color: #236fa1; text-decoration: underline;"><span style="font-size: 16px;"><strong>[Treating outliers]&nbsp;</strong></span></span></span><span style="font-size: 14px;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Method 2: F</span></strong></span><span style="color: #236fa1;"><span style="font-size: 16px;"><span style="color: #236fa1;"><span style="font-size: 16px;"><strong><u>looring and capping</u></strong></span></span><strong style="color: #236fa1; font-size: 16px;"> the outlier</strong></span></span></span></p>
<p><span style="color: #236fa1;"><span style="font-size: 16px;">In this technique, the outlier is capped at a certain value above the 90th percentile value or floored at a factor below the 10th percentile value.</span></span></p>

```python
# copy to a new dataframe
diabetes_df_demo2 = diabetes_df.copy()
```

```python
# calculate the 10th and 90th percentile of diabetes_df_demo2["BloodPressure"]
percentile_10 = np.percentile(diabetes_df_demo1["BloodPressure"], 10)
percentile_90 = np.percentile(diabetes_df_demo2["BloodPressure"],90)

# display the 10th and 90th percentile values
print("10th percentile is", percentile_10)
print("90th percentile is", percentile_90)
```

```python
# column that we are interested
x = "BloodPressure"

# capping the values below 10th percentile values
diabetes_df_demo2.loc[diabetes_df_demo2[x]< percentile_10, x] = percentile_10

# capping the values above 90th percentile values
diabetes_df_demo2.loc[diabetes_df_demo2[x]> percentile_90,x] = percentile_90
```

<p><span style="font-size: 12pt; color: #236fa1;">We plot the KDE plot on the diabetes_df_demo2["BloodPressure"] column, you will notice the distribution looks symmetric and the skew value is closed to zero.</span></p>

```python
# change the plotting style and figure size
sns.set(style = "ticks", color_codes = True, rc = {'figure.figsize':(5,5)})

# plot the KDE with color and fill
sns.kdeplot(diabetes_df_demo2["BloodPressure"], color = "brown", fill = True)

# find out the skewness of variable
skewness_0 = diabetes_df_demo2["BloodPressure"].skew()

# display the skewness value by add text annotation on plot
plt.text(40,0.036,"Skew:"+ str(skewness_0), size=16)
```

---

<p><span style="text-decoration: underline;"><span style="font-size: 14px;"><span style="color: #236fa1; text-decoration: underline;"><span style="font-size: 16px;"><strong>[Treating outliers]&nbsp;</strong></span></span></span><span style="font-size: 14px;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Method 3: Mean/median imputation</span></strong></span><span style="color: #236fa1;"><span style="font-size: 16px;"><strong>&nbsp;using IQR Score</strong></span></span></span></p>
<p><span style="color: #236fa1;"><span style="font-size: 16px;">As the mean value is highly influenced by the outliers, it is advised to replace the outliers with the mean/median value.</span></span></p>

```python
# copy to a new dataframe
diabetes_df_demo3 = diabetes_df.copy()
```

```python
# column that we are interested
x = 'BloodPressure'

# calculate the 25th and 75th percentile
q75, q25 = np.percentile(diabetes_df_demo3.loc[:,x],[75,25])

# calculate the inter-quartile
intr_qr = q75 + (1.5 * intr_qr)

# calculate the 1.5 times IQR above Q3 and below Q1
boxplot_max = q75 + (1.5 * intr_qr)
botplot_min = q25 - (1.5 * intr_qr)


# print the 1.5 times IQR above Q3 and below Q1
print('boxplot_max is',boxplot_max)
print('boxplot_min is', boxplot_min)
```

```python
# column that we are interested
x = 'BloodPressure'

# display the records below boxplot_min
diabetes_df_demo3.loc[diabetes_df_demo3[x] < boxplot_min,x]
```

```python
# column that we are interested
x = 'BloodPressure'

# display the records above boxplot_max
diabetes_df_demo3.loc[diabetes_df_demo3[x]>boxplot_max,x]
```

```python
# calculate the mean value of diabetes_df_demo3["BloodPressure"]
blood_pressure_mean = diabetes_df_demo3["BloodPressure"].mean()

# column that we are interested
x = "BloodPressure"

# replace records less than boxplot_min with mean value
diabetes_df_demo3.loc[diabetes_df_demo3[x]< boxplot_min, x] = blood_pressure_mean

# replace records greater than boxplot_max with mean value
diabetes_df_demo3.loc[diabetes_df_demo3[x]< boxplot_max, x] = blood_pressure_mean

```

<p><span style="color: #236fa1; font-size: 16px;">We plot the KDE plot on the diabetes_df_demo3["BloodPressure"] column, you will notice the distribution looks symmetric and the skew value is closed to zero.</span></p>

```python
# change the plotting style and figure size
sns.set(style = "ticks", color_codes = True, rc = {'figure.figsize':(5,5)})

# plot the KDE with color and fill
sns.kdeplot(diabetes_df_demo3["BloodPressure"], color = "brown", fill = True)

# find out the skewness of variable
skewness_0 = diabetes_df_demo3["BloodPressure"].skew()

# display the skewness value by add text annotation on plot
plt.text(40,0.046,"Skew:" + str(skewness_0),size=16)
```
