```python
# import matplotlib library
import matplotlib.pyplot as plt
import numpy as np

# sample data for this question
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

display(languages, popularity)

# color code for pie chart
colors = ["red", "orange", "yellow", "blue", "brown", "cyan"]
```

```python
# Q1
# write python code here
plt.figure(facecolor = 'white', figsize = (8,8))
plt.pie(x=popularity, labels=languages, autopct='%0.2f%%')
plt.title("Popularity of programming Language Worldwide, Oct 2017 compared to a year ago", size = 18)
plt.show()
```

<h2 id="Sample-Data"><span style="color: #e67e23;">Q3 Sample Data</span></h2>

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['Team 1', 'Team 2', 'Team 3', 'Team 4', 'Team 5']
men_means = [35, 23, 25, 25, 28]
women_means = [19, 30, 18, 20, 30]
```

```python
length = len(labels)
x = np.arange(length)
x
```

```python
# Q3
# write python code here
bar_width = 0.35
x = np.arange(len(labels))
fig = plt.figure(facecolor='white', figsize=(7, 7))
ax = fig.add_subplot(111)
ax.bar(x - bar_width/2, men_means, color='blue', width=bar_width, label='Men')
ax.bar(x + bar_width/2, women_means, color='purple', width=bar_width, label='Women')
ax.legend()
plt.title("Scores by team and gender", size=18)
plt.xlabel("List of teams", size=12)
plt.ylabel("Scores", size=12)
plt.xticks(x, labels)
plt.show()
```
