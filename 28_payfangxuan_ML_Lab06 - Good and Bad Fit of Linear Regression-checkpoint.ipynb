<h1 id="Good-Fit-of-Linear-Regression"><span style="color: #236fa1;">Good Fit of Linear Regression</span></h1>

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Load data</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">The following example code demonstrates a good fit for linear regression.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">First, we read the csv dataset file.</span></p>

```python
# create a dataframe from "linear_good_dataset.csv"
linear_good_df = pd.read_csv("./linear_good_dataset.csv")

# display linear_good_df
linear_good_df
```

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Scatter plot</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Plot the xData and yData using scatter plot to see how the data point scatter around the graph.</span></p>

```python
# change the plotting style and figure size
sns.set(style = "darkgrid", color_codes = True, rc = {'figure.figsize':(12,6)})

# plot the scatter plot of x1 and y1
sns.scatterplot(data = linear_good_df, x = "xData", y="yData")

# show the plot
plt.show()
```

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Heatmap on the correlation coefficient</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">We can create dataframe using the two variables and visualise the correlation of variables using Seaborn heatmap.</span></p>

```python
# change the plotting style
sns.set(style = "ticks", color_codes = True, rc = {'figure.figsize':(5,5)})

# using seaborn heatmap to plot the heatmap of correlation
sns.heatmap(linear_good_df.corr(method='pearson'), center = 0, annot = True)
```

<p><span style="font-size: 12pt; color: #236fa1;">From the heatmap above, w</span><span style="font-size: 12pt; color: #236fa1;">e observed a positive correlation coefficient of 0.92 between xData and yData.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">We use Scikit-Learn's LinearRegression estimator to fit this data and construct the best-fit line.</span></p>

---

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">X and Y data:</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Make the xData and yData columns into separate variables.</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #236fa1;">xData - independent variable</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #236fa1;">yData - dependent variable</span></p>

```python
# copy the x-axis data to x1
# reshape the train data into single element list
x1 = linear_good_df["xData"].array.reshape(-1,1)

# copy the y-axis data to y1
y1 = linear_good_df["yData"]
```

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Train the model - Linear regression</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">In the following Python code, we will be using LinearRegression from sklearn.linear_model package to train the linear model.</span></p>

```python
# importing the LinearRegression class from sklearn.linear_model
from sklearn.linear_model import LinearRegression

# create a new instance of the LinearRegression class and name it as lm
lm1 = LinearRegression()

# use the fit() method to train the linear regression model using x1 and y1
lm1.fit(x1,y1)
```

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Predict the dependent variable using the trained model</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Now we can use the trained model to predict the dependent variable.</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #236fa1;">First, we create 21 linear data points of the independent variables.&nbsp;</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #236fa1;">Next, we will use those data points to predict the dependent variables.</span></p>

```python
# create a list of 21 points in xfit
# predict the yfit values for each point of xfit
xfit1 = np.linspace(0,10,21)

# xfit1.reshape(-1, 1) - reshape the xfit1 data into single element list
yfit1 = lm1.predict(xfit1.reshape(-1,1))

# display the xfit1 and yfit1 in dataframe
table_disp = pd.DataFrame({"x1": xfit1, "y1 predicted": yfit1})
table_disp
```

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Evaluate - Calculate R-squared value</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Use sklearn's metrics to calculate the R-squared value.</span></p>

```python
# importing the LinearRegression class from sklearn.linear_model
from sklearn import metrics

# predict the dependent variable using the original x1 variable
y_pred = lm1.predict(x1)

# calculate and display the R-squared score
metrics.r2_score(y1, y_pred)
```

---

<h1 id="R2---R-squared-value"><span style="color: #236fa1;">[Recall] - R<sup>2</sup>&nbsp;value</span></h1>

<p><span style="font-size: 12pt; color: #236fa1;">R<span style="font-size: 12px; line-height: 0; position: relative; vertical-align: baseline; top: -0.5em;">2</span>, R-squared is a&nbsp;<strong>statistical measure of how close the data are to the fitted regression line</strong>.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">It is also known as the&nbsp;<strong>coefficient of determination</strong>, or the coefficient of multiple determination for multiple regression.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">The best possible score is 1.0, which can be negative (because the model can be arbitrarily worse). A constant model that always predicts the expected value of y, disregarding the input features, would get a score of 0.0.</span></p>
<p>&nbsp;</p>
<p><span style="font-size: 12pt; color: #236fa1;">The R-squared value is&nbsp;<strong>used to evaluate how the predicted line is well fitted</strong>&nbsp;to the data points.</span></p>

---

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Plot original and predicted data</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Now we have predicted data, we can start to plot the original data and predicted data together in a single plot.</span></p>

```python
# change the size of the plot figure
plt.figure(figsize = (8,6))

# creates a scatter plot with age on the x-axis and time_second on the y-axis
plt.scatter(x1,y1)

# creates a line plot with xfit on the x-axis and predicted yfit on the y-axis
plt.plot(xfit1,yfit1,color = 'magenta')

# change title and labels for plot
plt.title("Simulated data1")
plt.xlabel("x1")
plt.ylabel("y1")
# annotate r2 score on plot
r2_score = metrics.r2_score(y1,y_pred)
txt='R2 -' + str(r2_score)
plt.text(1,15,txt,fontsize = 15)
# add a marker of the intercept point
plt.plot(0,lm1.intercept_, ls = "", marker = "p", label = "points", color = 'orchid', markersize = 10)

# show the plot
plt.show()
```

<p><span style="text-decoration: underline; color: #236fa1;"><strong><span style="font-size: 12pt;">Slope&nbsp;coefficient and intercept&nbsp;</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">The slope coefficient and intercept of the predicted data can be display using coef_ and intercept_. We can observed the slope coefficient and intercept are close to 2 and -5.</span></p>

```python
print("Linear model slope:    ", lm1.coef_[0])
print("Linear model intercept:", lm1.intercept_)
```

<h1 id="Bad-fit-of-linear-regression"><span style="color: #236fa1;">Bad fit of linear regression</span></h1>

<p><span style="font-size: 12pt; color: #236fa1;">So far, we show data that almost best fitted to predict future values using linear regression.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Here, we demonstrate data that not fitted to use linear regression.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Let us create an example data where linear regression would not be the best method to predict future values.</span></p>

```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

%matplotlib inline
```

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Load data</span></strong></span></p>

```python
# create a dataframe from "linear_bad_dataset.csv"
linear2_df = pd.read_csv("./linear_bad_dataset.csv")

# display linear2_df
linear2_df
```

<p><span style="text-decoration: underline; color: #236fa1;"><strong><span style="font-size: 12pt;">Scatter plot</span></strong></span></p>
<p><span style="color: #236fa1;"><span style="font-size: 12pt;"><span style="color: #236fa1;"><span style="font-size: 16px;">Plot the xData and yData using scatter plot to see how the data point scatter around the graph.</span></span></span></span></p>

```python
# change the plotting style and figure size
sns.set(style = "darkgrid",color_codes = True, rc = {'figure.figsize':(12,6)})

# plot the scatter plot of x1 and y1
sns.scatterplot(data = linear2_df,x = "xData", y = "yData")

# show the plot
plt.show()
```

<p><span style="font-size: 12pt; color: #236fa1;">From the scatter plot above, we observed the data point are scattered around the plot with <strong>no correlation</strong>.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Typically, we will not continue with linear regression modelling. But for <strong>demonstration purposes</strong>, we will continue with normal linear regression modelling.</span></p>

<p><strong style="font-size: 14px;"><span style="font-size: 12pt; color: #236fa1; text-decoration-line: underline;">Heatmap on the correlation coefficient</span></strong></p>
<p><span style="color: #236fa1; font-size: 16px;">Next, we create dataframe of the two variables and visualise the correlation of variables using the Seaborn heatmap.</span></p>

```python
# change the plotting style
sns.set(style = "ticks", color_codes = True, rc = {'figure.figsize':(5,5)})

# using seaborn heatmap to plot the heatmap of correlation
sns.heatmap(linear2_df.corr(method='pearson'),center=0,annot = True)
```

<p><span style="font-size: 12pt; color: #236fa1;">From the heatmap above, that is another indication that both variables have no correlation at all.</span></p>

---

<p style="font-size: 14px;"><span style="text-decoration: underline;"><span style="text-decoration-line: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Train, predict and plot</span></strong></span></span></p>

```python
# copy the x-axis data to x2
# reshape the train data into single element list

x2 = linear2_df["xData"].array.reshape(-1,1)
# copy the y-axis data to y2
y2 = linear2_df["yData"]
```

```python
#importing the LinearRegression class from sklearn.linear_model
from sklearn.linear_model import LinearRegression

# create a new instance of the LinearRegression class and name it as lm
lm2 = LinearRegression()

# use the fit() method to train the linear regression model using x2 and y2
lm2.fit(x2,y2)
```

```python
# create a list of 20 points in xfit
# predict the yfit values for each point of xfit
xfit2 = np.linspace(0,110,31)
yfit2 = lm2.predict(xfit2[:, np.newaxis])
```

```python
# change the size of the plot figure
plt.figure(figsize = (8,6))

# creates a scatter plot with x2 on the x-axis and y2 on the y-axis
plt.scatter(x2,y2)

# creates a line plot with xfit on the x-axis and predicted yfit on the y-axis
plt.plot(xfit2,yfit2, color = 'red')

# change title and labels for plot
plt.title("Simulated data 2")
plt.xlabel("x2")
plt.ylabel("y2")
# show the plot
plt.show()
```

<p><span style="font-size: 12pt; color: #236fa1;">Now we can clearly see the slope coefficient of predicted line is very close to zero.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Thats means it is x2 is no relationship with y2.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">It tells us that this dataset is not suitable for linear regression to predict future values.</span></p>

---

<p><strong><span style="text-decoration: underline;"><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Find out R-squared value</span></span></strong></p>
<p><span style="font-size: 12pt; color: #236fa1;">Use sklearn's metrics to calculate the R-squared value.</span></p>

```python
# importing the LinearRegression class from sklearn.linear_model
from sklearn import metrics

# predict the dependent variable using the original x1 variable
y_pred = lm2.predict (x2)

# calculate and display the R-squared score
metrics.r2_score(y2,y_pred)
```

<p><span style="font-size: 12pt; color: #236fa1;">The R-squared value is almost zero, this indicate a bad fit of predicted line to original data.</span></p>
