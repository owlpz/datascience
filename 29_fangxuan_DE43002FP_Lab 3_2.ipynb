<h1 id="Scatter-plot"><span style="color: #e67e23;">Scatter plot</span></h1>

<p><span style="font-size: 12pt; color: #e67e23;">Import all relevant library files and create a dictionary with name and marks as given below.</span></p>

```python
# data to be plot
age = np.array([20, 30, 40, 50])
weight = np.array([45, 55, 52, 100])

# display the data
display(age, weight)
```

```python
# Set background color of the outer area of the plot and figure size
plt.figure(facecolor = 'white', figsize = (6,6))

# set title, on x & y axis labels
plt.title('Age vs Weight')
plt.xlabel('Age')
plt.ylabel('Weight')
# plot scatter
plt.scatter(age,weight, color='red',marker='D')

# turn off grid on bar graph
plt.grid(False)

# show bar graph
plt.show()
```

<h1 id="Boxplot"><span style="color: #e67e23;">Boxplot</span></h1>

<p><span style="font-size: 12pt; color: #e67e23;">Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum.</span></p>
<p><span style="font-size: 12pt; color: #e67e23;">Generate a box plot using random numbers:</span></p>

```python
# create a test data frame usng ramdom number generator
# this data frame have 5 columns = 'A', 'B', 'C', 'D', 'E'
test_df = pd.DataFrame(np.random.rand(10,5), columns=['A','B','C','D','E'])

# Set background color of the outer area of the plot and figure size
plt.figure(facecolor = 'white', figsize = (6,6))

# plot box plot
plt.boxplot(test_df)

# turn on grid on bar graph
plt.grid(False)

# show bar graph
plt.show()
```

<p><span style="font-size: 12pt; color: #e67e23;">A Box Plot is the visual representation of the statistical five-number summary of a given data set.</span></p>
<p><span style="font-size: 12pt; color: #e67e23;">A Five Number Summary includes:</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #e67e23;">- Minimum</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #e67e23;">- First Quartile</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #e67e23;">- Median (Second Quartile)</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #e67e23;">- Third Quartile</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #e67e23;">- Maximum</span></p>

<div>

</div>

```python
def get_box_plot_data(labels, bp):
    rows_list = []

    for i in range(len(labels)):
        dict1 = {}
        dict1['label'] = labels[i]
        dict1['lower_whisker'] = bp['whiskers'][i*2].get_ydata()[1]
        dict1['lower_quartile'] = bp['boxes'][i].get_ydata()[1]
        dict1['median'] = bp['medians'][i].get_ydata()[1]
        dict1['upper_quartile'] = bp['boxes'][i].get_ydata()[2]
        dict1['upper_whisker'] = bp['whiskers'][(i*2)+1].get_ydata()[1]
        rows_list.append(dict1)

    return pd.DataFrame(rows_list)
```

```python
attendance_percentage = [
    100, 100, 100, 100, 100, 100, 100, 100, 100, 88, 93, 98, 91, 100, 97, 94,
    100, 55, 68, 70, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100,
    100, 100, 100, 99, 98, 82, 100, 94, 93, 100, 82, 82, 88, 49, 68, 37, 100,
    92, 64, 100, 88, 100, 100, 100, 100, 100, 100, 100, 99, 97, 100, 100, 98,
    80, 85, 73, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 89, 94, 99,
    92, 88, 100, 89, 99, 98, 82, 100, 100, 97, 100, 98, 100, 100, 100, 100, 92,
    87, 100, 95, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100,
    100, 91, 100, 97, 100, 100, 100, 100, 100, 100, 100, 100, 95, 100, 91, 90,
    88, 100, 95, 99, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100,
    100, 100, 100, 99, 100
]

attendance_df = pd.DataFrame(attendance_percentage,
                             columns=['Attendance Rate'])

attendance_df.head()
```

```python
# set background color of the outer area of the plot and figure size
plt.figure(facecolor = 'white', figsize = (10,6))

# plot box plot
bp=plt.boxplot(attendance_df)

# turn on grid on bar graph
plt.grid(False)

# show bar graph
plt.show()
display(get_box_plot_data(attendance_df.columns, bp))

```
