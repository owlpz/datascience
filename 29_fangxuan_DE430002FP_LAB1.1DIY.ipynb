## Create dataframe from zip()

```python
# Create dataframe using Zip()

# create data list
name = ['Tommy', 'Kenny', 'James', 'Jinny', 'Ahmad']
gender = ['Male', 'Male', 'Male', 'Female', 'Male']
age = [25, 30, 26, 22, 23]
# contruct a zip
data_zip = set(zip(name,gender,age))

# display data_zip
data_zip
```

```python
# constructing dataframe from a zip
data_df3 = pd.DataFrame(data_zip, columns = ['name', 'gender', 'age'])

# display data_df3
data_df3
```
