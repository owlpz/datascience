<p><span style="font-size: 12pt; color: #236fa1;">In the statistical summary above, you notice the "Outcome" column was not listed. The "Outcome" column is categorical data and it is classified as a boolean. That's why it was not calculated and listed in the statistical summary.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Categorical data distribution statistics are useful in classification problems where we need to know the balance of categorical data.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">In case we have an imbalanced categorical data distribution (one class is having lots more records than another class). It may need special handling at the data preparation stage of the machine learning project.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">In Python, we use the groupby(), size(), reset_index() method to find out the categorical data distribution.</span></p>
<p><span class="fontstyle0" style="font-size: 12pt; color: #236fa1;">In the output below, it can be clearly seen that the number of observations with the outcome "FALSE" is almost double the number of observations with the outcome "TRUE".</span></p>

```python
# find out the distribution 
outcome_dist = diabetes_df.groupby("Outcome").size().reset_index(name = 'count')
outcome_dist
```

<p><span style="text-decoration: underline; font-size: 12pt; color: #236fa1;"><strong>Categorical data encoding</strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">We need to <strong>convert categorical data into an integer</strong>. One of the methods is easily applied <strong>one-hot encoding</strong> to this problem but that would <strong>simply increase the number of features</strong> and <strong>required higher computation power</strong>.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Instead, we derive a relation and map those categorical data.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">In Python, we can use <strong>map() method to do the categorical data encoding</strong>.</span></p>

```python
# find out the unique values in the "Outcome" column
unique_value = diabetes_df ["Outcome"].unique()
display(unique_value)
# map the unique values in the "Outcome" column to integer
diabetes_df['Outcome_int']=diabetes_df['Outcome'].map({unique_value[0]:1,unique_value[1]:0})

# display the records after mapping the categorical data to integer
# you will observe that TRUE mapped to 1, FALSE mapped to 0
diabetes_df.head()
```

<p><span style="text-decoration: underline; font-size: 12pt; color: #236fa1;"><strong>Drop columns from dataframe</strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">After categorical data encoding, we can remove the "Outcome" column.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">To drop specified labels from rows or columns, we can use drop() method.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Remove rows or columns by specifying label names and corresponding axis, or by specifying directly index or column names.&nbsp;</span></p>

```python
# drop the original column "Outcome"
diabetes_df = diabetes_df.drop("Outcome",axis=1)

# display the diabetes_df
diabetes_df
```

<p><span style="text-decoration: underline; font-size: 12pt; color: #236fa1;"><strong>Reviewing correlation between columns</strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">The <strong>relationship between two variables</strong> is called <strong>correlation</strong>.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">In statistics, the most common method for calculating correlation is <strong>Pearson&rsquo;s Correlation Coefficient</strong>.</span></p>
<p><br /><span style="font-size: 12pt; color: #236fa1;">It can have three values as follows:</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #236fa1;">1. <strong>Coefficient value = 1</strong>: It represents the <span style="text-decoration: underline;">strong positive correlation</span> between variables.</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #236fa1;">2. <strong>Coefficient value = -1</strong>: It represents the <span style="text-decoration: underline;">strong negative correlation</span> between variables.</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #236fa1;">3. <strong>Coefficient value = 0</strong>: It represents <span style="text-decoration: underline;">no correlation</span> at all between variables.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">It is always good to <strong>review the pairwise correlations of the variables in the dataframe</strong> before using them in the machine learning project because some machine learning algorithms such as linear regression and logistic regression will perform poorly if we have highly correlated variables.</span></p>
<p>&nbsp;</p>
<p><span style="font-size: 12pt; color: #236fa1;">In Python, <strong>corr() method is used to calculate a correlation matrix of dataframe</strong>.</span></p>

```python
# get the correlation of diabetes_df
# use the pearson correlation
diabetes_df.corr(method='pearson')
```

---

<p><span style="text-decoration: underline; font-size: 12pt; color: #236fa1;"><strong>Visualise the correlation of variables using the heatmap</strong></span></p>

<p><span style="text-decoration: underline; font-size: 12pt; color: #236fa1;"><strong>Skewed Distribution of variables</strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Skewness is defined as the distribution that is assumed to be Gaussian but appears distorted or shifted in one direction or another, or either to the left or right.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Reviewing the skewness of variables is one of the important tasks due to following reasons:</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #236fa1;">1. Presence of skewness in data requires the correction at the data preparation stage so that we can get more accuracy from our model.</span></p>
<p style="padding-left: 40px;"><span style="font-size: 12pt; color: #236fa1;">2. Most of the machine learning algorithms assumes that data has a Gaussian distribution i.e. either normal or bell-curved data.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">In Python, the skew() method is used to review the skewness of variables.</span></p>
<p>&nbsp;</p>
<p><span style="font-size: 12pt; color: #236fa1;">Note: in the output below, positive or negative skew can be observed. If the value is closer to zero, then it shows less skew.</span></p>
<p>&nbsp;</p>
<p><span style="font-size: 12pt; color: #236fa1;">More explanation: https://www.statisticshowto.com/probability-and-statistics/skewed-distribution/</span></p>

```python
# video explaining the skewed distribution
# source: https://www.youtube.com/watch?v=24YCewQ8s3M
from IPython.display import YouTubeVideo
YouTubeVideo('24YCewQ8s3M', width = 800, height = 500)
```

```python
# find out the statistical value of skewness
diabetes_df.skew()
```

<p><span style="text-decoration: underline; font-size: 12pt; color: #236fa1;"><strong>Skewed distribution explained</strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Symmetry distribution means that one half of the distribution is a mirror image of the other half.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">For example, the normal distribution is a symmetric distribution with no skew. The tails are exactly the same.</span></p>


<center>Normal distribution curve</center>

Source: https://www.statisticshowto.com/wp-content/uploads/2013/09/normal-distribution-probability.jpg

<p><span style="text-decoration: underline; font-size: 12pt; color: #236fa1;"><strong>Asymmetric distributions</strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">If one tail is longer than another, the distribution is skewed. These distributions are called asymmetric or asymmetrical distributions.&nbsp;</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">A <span style="text-decoration: underline;">left-skewed</span> distribution has a <strong>long left tail</strong>. Left-skewed distributions are also called negatively-skewed distributions. That&rsquo;s because there is a long tail in the negative direction on the number line. The mean is also to the left of the peak.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">A <span style="text-decoration: underline;">right-skewed</span> distribution has a <strong>long right tail</strong>. Right-skewed distributions are also called positive-skew distributions. That&rsquo;s because there is a long tail in the positive direction on the number line. The mean is also to the right of the peak.</span></p>


Source: https://www.statisticshowto.com/wp-content/uploads/2014/02/pearson-mode-skewness.jpg

<p><span style="text-decoration: underline; font-size: 12pt; color: #236fa1;"><strong>Kernel Density Estimate (KDE) plot using Seaborn</strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Kernel Density Estimate (KDE) Plot and Kdeplot is used to estimate the probability density function of the continuous or non-parametric from our data set curve in one or more dimensions it means we can create plot a single graph for multiple samples which helps in more efficient data visualization.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">kernel density estimation can plot univariate or bivariate distributions.</span></p>
<p>&nbsp;</p>
<p><span style="font-size: 12pt; color: #236fa1;">A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogue to a histogram.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">KDE represents the data using a continuous probability density curve in one or more dimensions.</span></p>
<p>&nbsp;</p>
<p><span style="font-size: 12pt; color: #236fa1;">Relative to a histogram, KDE can produce a plot that is less cluttered and more interpretable, especially when drawing multiple distributions. But it has the potential to introduce distortions if the underlying distribution is bounded or not smooth.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Like a histogram, the quality of the representation also depends on the selection of good smoothing parameters.</span></p>
<p>&nbsp;</p>
<p><span style="font-size: 12pt; color: #236fa1;">More explanation and example: https://seaborn.pydata.org/generated/seaborn.kdeplot.html</span></p>

```python
# change the plotting style and figure size
sns.set(style = "ticks", color_codes = True, rc = {'figure.figsize':(3,3)})

# plot the KDE with color and fill
sns.kdeplot(diabetes_df["Pregnancies"], color = "skyblue", fill = True)

# find out the skewness of variable
skewness_0 = diabetes_df["Pregnancies"].skew()

# display the skewness value by add text annotation on plot
plt.text(0,0.17,"Skew:" + str(skewness_0),size = 12)
```
