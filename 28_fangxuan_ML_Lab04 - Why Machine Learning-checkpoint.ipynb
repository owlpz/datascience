<h1 id="Example-of-Traditional-Programming"><span style="color: #236fa1;">Example of Traditional Programming</span></h1>

<h3><span style="color: #236fa1;">Examination mark grading system using traditional programming</span></h3>
<p><span style="font-size: 12pt; color: #236fa1;">We write a Python program that provides an examination grade based on a set of rules.</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">The grade will be awarded based on the mark according to table below:</span></p>

<div>

</div>

```python
# define function of exam grading rules
def exam_grading(mark):
    if mark <= 39:
        return 'F'
    elif mark <= 49:
        return 'E'
    elif mark <= 59:
        return 'D'
    elif mark <= 69:
        return 'C'
    elif mark <= 79:
        return 'B'
    elif mark <= 89:
        return 'A'
    else:
        return 'DIS'
```

```python
# student exam mark (0 - 100)
# you may edit the mark to test the program
student_mark = 85

# get the student grade using the exam_grading() function
student_grade = exam_grading(student_mark)

# print the student mark and grade
print('student mark is %d and grade is %s.' % (student_mark, student_grade))
```

<p><span style="font-size: 12pt;"><strong><span style="text-decoration: underline;"><span style="color: #236fa1; text-decoration: underline;">Unique values in the categorical data column</span></span></strong></span></p>
<p><span style="color: #236fa1; font-size: 12pt;">We need to find out the unique values in the categorical data column.</span></p>
<p><span style="color: #236fa1; font-size: 12pt;">In Python, we can use unique() or set() to display the unique values in the categorical data column.</span></p>

```python
# using unique() method, display the unique values in the categorical data column
score_df["Grade"].unique()
```

---

<p><span style="text-decoration: underline; color: #236fa1;"><strong><span style="font-size: 12pt;">Encode categorical data</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Now we know the unique values, we can encode these values to integers.</span></p>

```python
# encode the categorical data into integer using map()
score_df["Grade"] = score_df["Grade"].map({'DIS' :0,
                                         'A':1,
                                         'B':2,
                                         'C':3,
                                         'D':4,
                                         'E':5,
                                         'F':6})

# display score_df
score_df
```

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Note:</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">After the mapping, you will notice the score['Grade'] column is <strong>encoded from categorical data into an integer</strong>.</span></p>

---

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Input data and expected outcome</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">From score_df, we can create the train data from score_df['Score'] and the expected output data from score_df['Grade'].</span></p>

```python
# create the input data from score_df['Score']
# reshape the train data into single element list
x_train = score_df["Score"].array.reshape(-1,1)

# create the expected output data from score_df['Grade']
y_train = score_df['Grade']
```

---

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Train the data model using decision tree</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">We will use X_train and y_train to fit the data model using DecisionTreeClassifier().</span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Decision Tree is a classification machine learning algorithm. It classifies the grade based on the mark.</span></p>

```python
# import the DecisionTreeClassifier from scikit-learn package
from sklearn.tree import DecisionTreeClassifier

# DecisionTreeClassifier class
dtree = DecisionTreeClassifier(criterion = 'gini')

# fit the model using X_train and y_train
dtree.fit(x_train,y_train)
```

---

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Visualise the machine learning model</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">We can plot the flow chart of the data model created.</span></p>

```python
# import the plot_tree from scikit-learn package
from sklearn.tree import plot_tree

# define the feature name
fNames = ["Score"]

# define the grade based on the encoded categorical data
cNames = ['DIS','A','B','C','D','E','F']

# plot the flow chart
plt.figure(figsize = [6,4], dpi = 200)
plot_tree(dtree, feature_names = fNames, class_names = cNames, filled = True)
plt.show()
```

<p><span style="text-decoration: underline;"><strong><span style="font-size: 12pt; color: #236fa1; text-decoration: underline;">Get student grade</span></strong></span></p>
<p><span style="font-size: 12pt; color: #236fa1;">Get the student grades based on student marks using the data model created.</span></p>

```python
# student exam mark (0 - 100)
# you may edit the mark to test the program
student_mark_m1 = 78

# define the grade based on the encoded categorical data
cNames = ['DIS', 'A','B','C','D','E','F']

# get the grade from student mark
grade_predict = dtree.predict([[student_mark_m1],])

# map the grade based on the encoded categorical data
student_grade_m1 = cNames[grade_predict[0]]

# print the student mark and grade
print('ML Model: student mark is %d and grade is %s.'% (student_mark_m1,student_grade_m1))
```
